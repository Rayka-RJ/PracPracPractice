```python
# Dot-product

def matrix_dot_vector(a: list[list[float]], b: list[float]) -> list[int|float]:
    if any(len(a[i]) != len(b) for i in range(len(a))):
        return -1
    dot_product = [sum( b[j]*a[i][j] for j in range(len(b))) for i in range(len(a))]
    return dot_product

a = matrix_dot_vector([[1, 2, 3], [2, 4, 5], [6, 8, 9]], [1, 2, 3])
b = matrix_dot_vector([[1, 2], [2, 4], [6, 8], [12, 4]], [1, 2, 3])
c = matrix_dot_vector([[1.5, 2.5], [3.0, 4.0]], [2, 1])
print(a,b,c)
```

```python
# transpose

def transpose_matrix(a: list[list[int|float]]) -> list[list[int|float]]:
	b = [[a[j][i] for j in range(len(a))] for i in range(len(a[0]))]
	return b

a = transpose_matrix([[1,2,3], [4,5,6]])
b = transpose_matrix([[1,2], [3,4],[5,6], [7,8]]) 
print(a,b)
```

```python
# Reshape matrix
def reshape_matrix(a:list[list[int|float]], new_shape: tuple[int, int]) -> list[list[int|float]]:
    if len(a)*len(a[0]) != new_shape[0] * new_shape[1]:
        return []
    return np.array(a).reshape(new_shape[0], new_shape[1]).tolist()

a = reshape_matrix([[1,2,3,4], [5,6,7,8]], [4,2])
print(a)
```

```python
# Calculate Mean by Row or Column
def calculate_matrix_mean(matrix: list[list[float]], mode:str) -> list[float]:
    if mode == 'row':
        return [ sum(row)/len(row) for row in matrix]
    elif mode == 'column':
        return [ sum([matrix[j][i] for j in range(len(matrix))])/len(matrix[0]) for i in range(len(matrix[0])) ]

a = calculate_matrix_mean( [[1, 2, 3], [4, 5, 6], [7, 8, 9]], 'column')
print(a)
```
