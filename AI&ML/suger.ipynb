```python
import numpy as np
```

```python
''' 
@
'''
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])

C = A @ B  # 矩阵乘法
print(C)
```

```python
# 语法糖

def my_decorator(func):
    def wrapper():
        print("✨ Before function")
        func()
        print("✨ After function")
    return wrapper

@my_decorator
def say_hello():
    print("Hello, Kaka!")

say_hello()
```

```python
'''
*
'''

# 函数打包
def greet(*names):
    for name in names:
        print(f"Hello, {name}!")

greet("Kaka", "Ray", "Zoe")

# 解包
args = [1, 2, 3]
print(*args)  # 相当于 print(1, 2, 3)

# 结合
# *a 是在 “解包 iterable”（可迭代对象） —— 把 a 的元素一个个拿出来，展开到 b 中。但是shallow copy
a = [1, 2]
b = [*a, 3, 4]
print(b)  # 输出: [1, 2, 3, 4]
```
