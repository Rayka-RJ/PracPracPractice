```python
# Data Structure Preparation

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def createLinkedList(nums: list):
    if not nums:
        return None
    
    head = ListNode(nums[0])
    current = head

    for num in nums[1:]:
        current.next = ListNode(num)
        current = current.next

    return head

def printLinkedList(head):
    current = head
    while current:
        print(current.val, end = ' -> ')
        current = current.next
    print(None)
```

```python
# 203. Remove Linked List Elements

class Solution:
    def removeElements(self, head, val: int):
        dummy = ListNode(next=head)
        current = dummy
        
        while current.next:
            if current.next.val == val:
                current.next = current.next.next
            else:
                current = current.next

        return dummy.next
```

```python
head = createLinkedList([1,2,6,3,4,5,6])
val = 6

a = Solution()
result = a.removeElements(head,val)
printLinkedList(result)
```

```python
# 206. Reverse Linked List
class Solution:
    def reverseList(self, head):
        prev = None
        current = head
        while current:
            temp = current.next # save the next node
            current.next = prev # reverse
            prev = current # update
            current = temp # update
            
        return prev
```

```python
head = createLinkedList([1,2,3,4,5,6])

a = Solution()
result = a.reverseList(head)
printLinkedList(result)
```

```python
# 24. Swap Nodes in Pairs
class Solution:
    def swapPairs(self, head):
        dummy = ListNode(0,next=head)
        prev = dummy
        current = dummy.next
        while current and current.next:
            # take
            temp1 = current
            temp2 = current.next

            # operation
            temp1.next = temp2.next
            temp2.next = temp1
            prev.next = temp2

            # move
            prev = current
            current = current.next
            
        return dummy.next
```

```python
head = createLinkedList([1,2,3,4])

a = Solution()
result = a.swapPairs(head)
printLinkedList(result)
```

```python
# 19. Remove Nth Node From End of List

class Solution:
    def removeNthFromEnd(self, head, n: int):
        dummy = ListNode(next=head)
        front = head
        current = dummy

        for _ in range(n):
            front = front.next
        
        while front:
            front = front.next
            current = current.next
        
        current.next = current.next.next
        return dummy.next
```

```python
head = createLinkedList([1,2,3,4,5])
n = 2

a = Solution()
result = a.removeNthFromEnd(head, n)
printLinkedList(result)
```

```python
# 707. Design Linked List

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class MyLinkedList:

    def __init__(self):
        self.dummy = ListNode()
        self.length = 0

    def get(self, index: int) -> int:
        if index < 0 or index >= self.length:
            return -1
        current = self.dummy.next
        for _ in range(index):
            current = current.next
        return current.val

    def addAtHead(self, val: int) -> None:
        self.dummy.next = ListNode(val,next=self.dummy.next)
        self.length += 1

    def addAtTail(self, val: int) -> None:
        current = self.dummy
        while current.next:
            current = current.next
        current.next = ListNode(val)
        self.length += 1

    def addAtIndex(self, index: int, val: int) -> None:
        if index < 0 or index > self.length:
            return
        
        current = self.dummy
        for _ in range(index):
            current = current.next
        current.next = ListNode(val, current.next)
        self.length += 1

    def deleteAtIndex(self, index: int) -> None:
        if index < 0 or index >= self.length:
            return

        current = self.dummy
        for _ in range(index):
            current = current.next
        
        current.next = current.next.next
        self.length -= 1
```

```python
# 160. Intersection of Two Linked Lists

class Solution:
    def getIntersectionNode(self, headA, headB):
        sz1, sz2 = 0,0

        current = headA
        while current:
            current = current.next
            sz1 += 1
        
        current = headB
        while current:
            current = current.next
            sz2 += 1

        if sz1 > sz2:
            l, s = headA, headB
        else:
            l, s = headB, headA
        
        for _ in range(abs(sz1-sz2)):
            l = l.next
        
        while l:
            if l == s:
                return l
            l = l.next
            s = s.next
```

```python
# 142. Linked List Cycle II

class Solution:
    def detectCycle(self, head):
        fast, slow = head, head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if fast == slow:
                slow = head
                while slow != fast:
                    slow = slow.next
                    fast = fast.next
                return slow
        return None
```
