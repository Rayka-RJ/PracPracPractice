```python
## 242. Valid Anagram

class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        char1, char2 = dict(), dict()
        for w in s:
            if w not in char1:
                char1[w] = 1
            else:
                char1[w] += 1
        for j in t:
            if j not in char2:
                char2[j] = 1
            else:
                char2[j] += 1
        return char1 == char2
```

```python
# 1002. Find Common Characters
class Solution:
    def commonChars(self, words):
        # 26 alphabet

        def hashList(word):
            res = {chr(i): 0 for i in range(97, 123)}
            for i in word:
                if i in res:
                    res[i] += 1
            return res
        
        record = hashList(words[0])
        for word in words[1:]:
            temp = hashList(word)
            for k in record.keys():
                record[k] = min(temp[k], record[k])

        res = []
        for k,v in record.items():
            if v != 0:
                for _ in range(v):
                  res.append(k)
        return res
```

```python
word = ["cool","lock","cook"]

a = Solution()
print(a.commonChars(word))
```

```python
# 349. Intersection of Two Arrays

class Solution:
    def intersection(self, nums1, nums2):
        tbl = {}
        for num in nums1:
            tbl[num] = tbl.get(num, 0) + 1
        
        res = set()
        for num in nums2:
            if num in tbl:
                res.add(num)
                del tbl[num]
        
        return list(res)
```

```python
ls1 = [1,2,2,1]
ls2 = [2,2]

a = Solution()
print(a.intersection(ls1,ls2))
```

```python
# 202. Happy Number
class Solution:
    def isHappy(self, n: int) -> bool:
        rec = set()
        rec.add(n)

        def loop(n):
            dig = []
            while n != 0:
                dig.append(n%10)
                n //= 10
            res = sum(map(lambda x: x*x, dig))
            return res

        while True:
            res = loop(n)
            if res != 1 and res in rec:
                return False
            elif res == 1:
                return True
            rec.add(res)
            n = res
```

```python
n = 19
a = Solution()
print(a.isHappy(n))
```

```python
# 1. Two Sum
class Solution:
    def twoSum(self, nums, target):
        rec = {}
        for i in nums:
            rec[i] = rec.get(i,0) + 1
        
        for k,v in rec.items():
            rec[k] -= 1
            if target - k == k and rec[k] != 0:
                return [i for i,v in enumerate(nums) if v == k]
            if target - k in rec and rec[target - k] != 0:
                return [nums.index(k), nums.index(target - k)]

```

```python
# 454. 4Sum II
class Solution:
    def fourSumCount(self, nums1, nums2, nums3, nums4):

        def twoSum(nums1, nums2):
            res = {}
            for i in nums1:
                for j in nums2:
                    s = i + j
                    res[s] = res.get(s, 0) + 1
            return res
        
        ab = twoSum(nums1, nums2)
        cd = twoSum(nums3, nums4)

        count = 0
        for k,v in ab.items():
            if 0 - k in cd and cd[0-k] != 0:
                count += v*cd[0-k]
        return count
```

```python
nums1 = [-1,-1]
nums2 = [-1,1]
nums3 = [-1,1]
nums4 = [1,-1]

a = Solution()
print(a.fourSumCount(nums1, nums2, nums3, nums4))
```

```python
# 383. Ransom Note
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str):

        def hashset(lst):
            a = {}
            for i in lst:
                a[i] = a.get(i,0) + 1
            return a

        a = hashset(ransomNote)
        b = hashset(magazine)
        count = 0
        for k,v in a.items():
            if k in b and b[k] >= v:
                count += 1
        return count == len(a)
    
    # better one
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        if len(magazine) < len(ransomNote):
            return False
        
        for r in ransomNote:
            if r not in magazine:
                return False
            ransomNote = ransomNote.replace(r, '', 1)
            magazine = magazine.replace(r, '', 1)
        
        return True
```

```python
ransomNote = "bg"
magazine = "efjbdfbdgfjhhaiigfhbaejahgfbbgbjagbddfgdiaigdadhcfcj"
a = Solution()
print(a.canConstruct(ransomNote, magazine))
```

```python
# 15.3Sum

'''
This is a time-expired methods
'''
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        res = set()
        sing, doub = dict(), set()
        for num in nums:
            sing[num] = sing.get(num, 0)+1
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if (nums[i], nums[j]) not in doub and (nums[j], nums[i]) not in doub:
                    doub.add((nums[i], nums[j]))

        for pair in doub:
            target = -(pair[0] + pair[1])
            temp_dic = sing.copy()
            if sing.get(target, 0) > 0:
                temp_dic[pair[0]] -= 1
                temp_dic[pair[1]] -= 1
                if temp_dic.get(target, 0) > 0:
                    temp = sorted([pair[0], pair[1], target])
                    res.add(tuple(temp))

        return [list(i) for i in res]
```

```python
# 15.3Sum
'''
two pointers:
- step 1: set the first num
- step 2: for rest list slice, two pointers to find the 0
- step 3: skip all the same value 
'''
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        res = []

        for i, num in enumerate(nums):
            if num > 0:
                break
            if i > 0 and nums[i-1] == num:
                continue

            a = nums[i+1:]
            j, k = 0, len(a) - 1
            while j<k:
                b = a[j] + a[k]
                if b > -nums[i]:
                    k -= 1
                elif b < -nums[i]:
                    j += 1
                else:
                    res.append([nums[i], a[j], a[k]])
                    j += 1
                    k -= 1
                    while a[j] == a[j-1] and j<k:
                        j += 1
        
        return res
```

```python
nums = [-1,0,1,2,-1,-4,-2,-3,3,0,4]
a = Solution()
print(a.threeSum(nums))
```
