```python
# 1. Two Sum
class Solution:
    def twoSum(self, nums, target):
        rec = {}
        for i, num in enumerate(nums):
            remain = target-num
            if remain in rec:
                return [rec[remain],i]
            rec[num] = i

nums = [3,2,4]
a = Solution()
res = a.twoSum(nums,6)
res
```

```python
# 41. First Missing Positive
class Solution:
    def firstMissingPositive(self, nums: List[int]) -> int:
        # step1 - 清理
        for i in range(len(nums)):
            if nums[i] <= 0 or nums[i] > len(nums):
                nums[i] = len(nums) + 1
        
        # step2 - 标记
        # 注意从idx[0,i]到value[1,i+1]的边界表示方法，需要-1来标记
        # 用idx代替value的话，即如果该value在数组里，就标记相对应idx的值，比如1在数据里，就标记idx=0的值，此时就需要value-1
        for i in range(len(nums)):
            v = abs(nums[i])
            if 0 < v <= len(nums):
                nums[v-1] = -abs(nums[v-1])

        # step3 - 寻找
        for i in range(len(nums)):
            if nums[i] > 0:
                return i + 1
    
        return len(nums) + 1
nums = [1,2,3,0,7]
a = Solution()
res = a.firstMissingPositive(nums)
res
```

```python
# 49. Group Anagrams
class Solution:
    def groupAnagrams(self, strs: List[str]):  
        dic = defaultdict(list)
        for s in strs:
            key = ''.join(sorted(s))
            dic[key].append(s)
        return list(dic.values())

strs = ["eat","tea","tan","ate","nat","bat"]
a = Solution()
res = a.groupAnagrams(strs)
res
```

```python
# 128. Longest Consecutive Sequence
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        no_dup = set(nums)
        long = 1
        for num in no_dup:
            if num - 1 not in no_dup:
                cmt = 1
                while num + cmt in no_dup:
                    cmt += 1
                long = max(long, cmt)

        return long
    
nums = [100,4,200,1,3,2]
a = Solution()
res = a.longestConsecutive(nums)
res
```

```python
# 208. Implement Trie (Prefix Tree)
# 构造题心法：
# - 构造题先定“最小必要信息”，
# - 再统一一条“走路骨架”，
# - 最后只改“分叉条件”。
class Node:
    def __init__(self) -> None:
        self.children = dict()
        self.is_end = False

class Trie:

    def __init__(self):
        self.root = Node()

    def insert(self, word: str) -> None:
        cur = self.root
        for c in word:
            if c not in cur.children:
                cur.children[c] = Node()
            cur = cur.children[c]
        cur.is_end = True

    def search(self, word: str) -> bool:
        cur = self.root
        for c in word:
            if c not in cur.children:
                return False
            cur = cur.children[c]
        return cur.is_end

    def startsWith(self, prefix: str) -> bool:
        cur = self.root
        for c in prefix:
            if c not in cur.children:
                return False
            cur = cur.children[c]
        return True


# Your Trie object will be instantiated and called as such:
word = 'apple'
prefix = 'app'
obj = Trie()
obj.insert(word)
param_2 = obj.search(word)
param_3 = obj.startsWith(prefix)
param_3
```

```python
# 560.Subarray Sum Equals K
'''
需要考虑重复，截取一定的子集区域，要保持连贯（不能sorted，要保持原始位置）。
前缀和！
'''
class Solution:
    def subarraySum(self, nums: List[int], k: int):
        pair = 0
        # prefix sum
        prefix = 0
        check = {0:1}
        for num in nums:
            prefix += num
            if prefix-k in check:
                pair += check[prefix-k]
            check[prefix] = check.get(prefix, 0) + 1
        return pair

nums, k = [1,1,1], 2
a = Solution()
res = a.subarraySum(nums,k)
res
```
