```python
from typing import List, Dict, Tuple, Optional
from collections import deque
import heapq
```

```python
class Solution:
    def isValid(self, s: str) -> bool:
        paren = {'(':')', '{':'}', '[':']'}
        store = []
        for p in s:
            if p in paren:
                store.append(p)
            else:
                if not store or p != paren[store.pop()]:
                    return False
        return not store

s = "(())"
a = Solution()
res = a.isValid(s)
res
```
