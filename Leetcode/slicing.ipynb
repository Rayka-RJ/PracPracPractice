```python
# 3. Longest Substring Without Repeating Characters
# 稳定检查是否有重复，一个固定指针（for-loop）一个动态指针（while-loop）
# 注意先分window，再内部检查duplicates

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        if not s: return 0
        res = 1
        for i in range(len(s)):
            check = set()
            check.add(s[i])
            cnt = 1
            j = i + 1
            while j < len(s):
                if s[j] not in check:
                    check.add(s[j])
                    cnt += 1
                    j += 1
                else:
                    res = max(res, cnt)
                    break
            res = max(res, cnt)
        return res

s = "au"
a = Solution()
res = a.lengthOfLongestSubstring(s)
res
```

```python

# 31. Next permutation

class Solution:
    def nextPermutation(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        i, j = len(nums)-2, len(nums)-1
        # 找左
        while i >= 0 and nums[i] >= nums[i+1]:
            i -= 1
        if i < 0:
            nums.reverse()
            return

        # 找右
        while j >= 0 and nums[j] < nums[i]:
                j -= 1
        
        # swap
        nums[i], nums[j] = nums[j], nums[i]
        # 逆序
        l,r = i + 1, len(nums)-1
        while l < r:
            nums[l], nums[r] = nums[r], nums[l]
            l += 1
            r -= 1

nums = [1,2,3]
a = Solution()
res = a.nextPermutation(nums)
nums
```

```python
# 76. Minimum Window Substring
# 好复杂一题目

from math import inf
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        # 1/统计频次
        need = dict()
        for i in t:
            if i not in need:
                need[i] = 1
            else:
                need[i] += 1
        
        # 2/初始化
        miss = len(t)
        left = 0
        best_len, best_left = float('inf'), 0

        # 3/right遍历，添加窗口
        for r, ch in enumerate(s):
            if ch in need:
                need[ch] -= 1
                if need[ch] >= 0:
                    miss -= 1
        
        # 4/字符找全，left收缩窗口
            while miss == 0:
                # 4.1 如果满足边界，更新left和长度
                win_len = r - left + 1
                if win_len < best_len:
                    best_len = win_len
                    best_left = left

                # 4.2 尝试再次移动left
                c = s[left]
                if c in need:
                    need[c] += 1 # 可能有冗余元素
                    # 移到大动脉，不满足条件
                    if need[c] > 0:
                        miss += 1
                left += 1
        if best_len == float('inf'):
            return ''
        
        return s[best_left:best_left+best_len]

s = "ADOBECODEBANC"
t = "ABC"
a = Solution()
res = a.minWindow(s,t)
res
```

```python
# 239. Sliding Window Maximum
# deque存索引，一边控制窗口长度一边比大小

from collections import deque

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        res = []
        dq = deque()
        for r in range(len(nums)):
            l = r - k + 1
            # window
            if dq and dq[0] < l:
                dq.popleft()
            
            # compare
            while dq and nums[dq[-1]] <= nums[r]:
                dq.pop()
            dq.append(r)
            
            # record
            if r >= k - 1:
                res.append(nums[dq[0]])
        return res

nums = [1,3,-1,-3,5,3,6,7]
k = 3
a = Solution()
res = a.maxSlidingWindow(nums, k)
res
```

```python
# 438. Find All Anagrams in a String

class Solution:
    def findAnagrams_exceed_time_limit(self, s: str, p: str) -> List[int]:
        # 1/ 记录
        check = dict()
        for i in p:
            check[i] = check.get(i, 0) + 1
        
        res = []
        # 2/ 遍历
        for i in range(len(s)-len(p)+1):
            c = check.copy()
            if s[i] not in c: continue
            j = i
            while j < i + len(p):
                if s[j] not in c or c[s[j]] <= 0:
                    break
                c[s[j]] -= 1
                j += 1
            if j == i + len(p):
                res.append(i)
        
        return res

# 使用一个特殊的计数结构
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        # 1/ 字典计数
        need = dict()
        for i in p:
            need[i] = need.get(i, 0) + 1
        
        miss = len(p)
        left = 0
        res = []

        # 2/ 遍历
        for right, ch in enumerate(s):
            if ch in need:
                need[ch] -= 1
                if need[ch] >= 0:
                    miss -= 1
            
            # 3/ 移动左边
            if right - left + 1 > len(p):
                out = s[left]
                if out in need:
                    need[out] += 1
                    if need[out] > 0:
                        miss += 1
                left += 1

            if right - left + 1 == len(p) and miss == 0:
                res.append(left)
        
        return res

s = "cbaebabacd"
p = "abc"
a = Solution()
res = a.findAnagrams(s,p)
res
```
