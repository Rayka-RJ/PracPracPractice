```python
# 3. Longest Substring Without Repeating Characters
# 稳定检查是否有重复，一个固定指针（for-loop）一个动态指针（while-loop）
# 注意先分window，再内部检查duplicates

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        if not s: return 0
        res = 1
        for i in range(len(s)):
            check = set()
            check.add(s[i])
            cnt = 1
            j = i + 1
            while j < len(s):
                if s[j] not in check:
                    check.add(s[j])
                    cnt += 1
                    j += 1
                else:
                    res = max(res, cnt)
                    break
            res = max(res, cnt)
        return res

s = "au"
a = Solution()
res = a.lengthOfLongestSubstring(s)
res
```

```python
# 76. Minimum Window Substring
# 好复杂一题目

from math import inf
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        # 1/统计频次
        need = dict()
        for i in t:
            if i not in need:
                need[i] = 1
            else:
                need[i] += 1
        
        # 2/初始化
        miss = len(t)
        left = 0
        best_len, best_left = float('inf'), 0

        # 3/right遍历，添加窗口
        for r, ch in enumerate(s):
            if ch in need:
                need[ch] -= 1
                if need[ch] >= 0:
                    miss -= 1
        
        # 4/字符找全，left收缩窗口
            while miss == 0:
                # 4.1 如果满足边界，更新left和长度
                win_len = r - left + 1
                if win_len < best_len:
                    best_len = win_len
                    best_left = left

                # 4.2 尝试再次移动left
                c = s[left]
                if c in need:
                    need[c] += 1 # 可能有冗余元素
                    # 移到大动脉，不满足条件
                    if need[c] > 0:
                        miss += 1
                left += 1
        if best_len == float('inf'):
            return ''
        
        return s[best_left:best_left+best_len]

s = "ADOBECODEBANC"
t = "ABC"
a = Solution()
res = a.minWindow(s,t)
res
```
