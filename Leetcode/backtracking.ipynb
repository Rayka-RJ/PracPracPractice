### DFS 和 Backtracking 的关系
DFS（深度优先搜索） 也是一种递归搜索方法，它按照一个路径一直深入，直到到达叶子节点或终止条件。
Backtracking 其实是 DFS 的一种特殊形式，它在遍历的过程中：
有明确的约束条件（如剪枝）
需要撤销错误的选择（即回溯）
通俗理解：

DFS 更像是“走到底”再回溯”
Backtracking 是“有策略的 DFS”，它会在不满足条件时提前剪枝

### 总结：Backtracking vs DFS vs BFS
算法 | 数据结构	| 适用场景	| 特点|
-----|-----|-----|-----|
DFS（深度优先搜索）|	递归 / 栈|	图搜索、路径搜索|	走到底再回溯，不剪枝|
BFS（广度优先搜索）|	队列	|最短路径	|按层遍历，不回溯|
Backtracking（回溯）|	递归 / 栈	|组合、排列、子集	|DFS + 剪枝 + 回溯|

- 什么时候用 Backtracking？
所有解都需要被枚举时（如排列、组合、子集）。
问题涉及递归 + 试探（搜索）+ 剪枝。
图搜索时如果不需要找最短路径，用 DFS/回溯比 BFS 更高效。
- 什么时候用 DFS？
图搜索（如连通性问题、拓扑排序）。
找单一路径（如迷宫问题、路径搜索）。
- 什么时候用 BFS？
找最短路径（如二分图、最短单源路径）。
所有路径长度相近，不需要回溯。


### 结论:
1. Backtracking ≈ DFS + 剪枝 + 回溯
2. DFS 适用于路径搜索，Backtracking 适用于排列/组合
3. BFS 不涉及回溯，适用于最短路径问题
不同问题需要不同搜索策略，选择合适的方法能提高效率！
