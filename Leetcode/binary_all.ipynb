```python
# 4.Median of Two Sorted Arrays
'''
- 这道题是要求我们在 两个有序数组中找到它们的中位数。
- 输入是两个递增排序的整型数组 nums1 和 nums2，输出是它们合并后数组的中位数。
- Naive解法是将两个数组合并，然后排序，再找中位数。时间复杂度是 O((m+n)log(m+n))，因为有一个排序过程；空间复杂度是 O(m+n)，因为创建了新数组。
- 我使用了Binary Search, 时间复杂度是 O(log(min(m, n)))。具体做法是：
    - 保证 nums1 比 nums2 短，然后在 nums1 上进行二分；
    - 设定一个分割点，把两个数组一共划分成左右两部分，使得左边的最大值小于等于右边的最小值；
    - 最终根据数组总长度是奇数还是偶数，返回中位数。
- 这种方法的空间复杂度是 O(1)，因为只用了常数级变量，没有新建数组。
- 这道题的关键是转换思路，不直接合并数组，而是通过在短数组上二分寻找一个合理的划分位置。
'''

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        if len(nums1) > len(nums2):
            nums1, nums2 = nums2, nums1
        left, right = 0, len(nums1)
        while left <= right:
            i = (left + right)//2
            nums1_left_max = float('-inf') if i == 0 else nums1[i-1]
            nums1_right_min = float('inf') if i == len(nums1) else nums1[i]

            j = (len(nums1)+len(nums2)+1)//2 - i
            nums2_left_max = float('-inf') if j == 0 else nums2[j-1]
            nums2_right_min = float('inf') if j == len(nums2) else nums2[j]

            if nums1_left_max <= nums2_right_min and nums2_left_max <= nums1_right_min:
                if (len(nums1)+len(nums2)) %2 == 0:
                    return (max(nums1_left_max, nums2_left_max) + min(nums1_right_min, nums2_right_min))/2
                else:
                    return max(nums1_left_max, nums2_left_max)
            elif nums1_left_max > nums2_right_min:
                right -= 1
            else:
                left += 1

nums1 = [1,3] 
nums2 = [2]
a = Solution()
res = a.findMedianSortedArrays(nums1, nums2)
res
```

```python
# 15.3Sum
'''
- 这道题要求我们找出数组中所有不重复的三元组，使得这三个数的和为 0。
- 输入是一个整数数组 nums，输出是一个二维数组，包含所有满足条件的不重复三元组。
- 最简单的做法是三重循环枚举所有三元组，判断是否为 0。时间复杂度是 O(n³)，空间复杂度取决于去重方式，一般是 O(n)。
- 我使用了 排序 + 双指针 的方法，把时间复杂度优化到 O(n²)。具体做法是：
    - 先对数组排序；
    - 固定一个数 num，然后在其右侧用左右指针找两个数，使得三者之和为 0；
    - 为了去重，跳过重复的固定数，指针移动时也跳过重复的元素。
    - 这种方法的空间复杂度是 O(1)（不计输出结果）。
- 这道题的核心是将暴力三重循环优化为 排序 + 双指针 的双层结构，同时注意 如何跳过重复元素 保证不出现重复解。
'''

class Solution:
    def threeSum(self, nums):
        nums.sort()
        res = []
        for i, num in enumerate(nums):
            if i > 0 and num == nums[i-1]:
                continue
            two_pointer = nums[i+1:]
            j,k = 0, len(two_pointer)-1
            while j < k:
                two_sum = two_pointer[j] + two_pointer[k]
                if two_sum + num > 0:
                    k -= 1
                elif two_sum + num < 0:
                    j += 1
                else:
                    res.append([two_pointer[j], two_pointer[k], num])
                    k -= 1
                    j += 1
                    while j < k and two_pointer[j] == two_pointer[j-1]:
                        j += 1
        return res

nums = [-1,0,1,2,-1,-4]
a = Solution()
res = a.threeSum(nums)
res
```

```python
# 18. 4Sum
'''
- 这道题是要求我们找出所有不重复的四元组，使得它们的和等于目标值 target。
- 输入是一个整数数组 nums 和一个目标值 target，输出是一个二维数组，包含所有满足条件的不重复四元组。
- 最朴素的解法是使用四重循环枚举所有组合，判断是否满足条件。
- 时间复杂度是 O(n⁴)，效率很低，不适合大数据量。
- 我使用了 排序 + 双重 for 循环 + 双指针 的优化方法，把时间复杂度降到 O(n³)。具体思路是：
    - 先对数组排序；
    - 枚举前两个数 nums[i] 和 nums[j]，然后对后面的部分用双指针寻找剩下两个数；
    - 每次找到一个组合后，要跳过重复元素，避免重复解。
    - 这个方法的空间复杂度是 O(1)（除了结果集以外）。
- 这道题的关键是将四层暴力循环优化为 两层循环 + 双指针，同时通过排序和跳过重复元素来保证去重正确且高效。
'''
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        res = []
        for i in range(len(nums)):
            if i > 0 and nums[i] == nums[i-1]:continue

            for j in range(i+1, len(nums)):
                if j > i+1 and nums[j] == nums[j-1]: continue

                left, right = j+1, len(nums)-1
                while left < right:
                    sum_all = nums[i] + nums[j] + nums[left] + nums[right]
                    if sum_all > target:
                        right -= 1
                    elif sum_all < target:
                        left += 1
                    else:
                        res.append([nums[i], nums[j], nums[left], nums[right]])
                        left += 1
                        while left < right and nums[left] == nums[left-1]:
                            left += 1
                        right -= 1
                        while left < right and nums[right] == nums[right+1]:
                            right -= 1

        return res

nums, target = [1,-2,-5,-4,-3,3,3,5], -11
a = Solution()
res = a.fourSum(nums, target)
res
```

```python
# 33. Search in Rotated Sorted Array
''' 
- 这道题要求我们在一个旋转过的升序数组中查找一个目标值，如果找到返回下标，否则返回 -1。
- 输入是一个整数数组 nums（无重复，原本升序，经过一次旋转）和一个整数 target，输出是目标值的下标或 -1。
- 最简单的方法是遍历整个数组查找目标值。
- 时间复杂度是 O(n)，空间复杂度是 O(1)，但没有利用有序性质。
- 我使用了一个变种的 二分查找，时间复杂度优化到了 O(log n)。思路是：
    - 每次比较中间元素 nums[mid] 和目标值；
    - 通过判断哪一边是有序的（nums[left] <= nums[mid] 或反之），
    - 再判断 target 是否落在这个有序段中，从而决定向哪一边收缩搜索范围。
    - 这种方法空间复杂度是 O(1)，只使用了几个指针变量。
- 这道题的关键是理解旋转数组的结构特点，并且在二分查找中判断哪一段是有序的，从而排除一半搜索区间。
'''

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left, right = 0 , len(nums)-1
        while left <= right:
            mid = (left + right)//2
            if nums[mid] == target:
                return mid

            # left is sorted
            if nums[left] <= nums[mid]:
                if nums[left] <= target < nums[mid]:
                    right = mid - 1
                else:
                    left = mid + 1
            # right is sorted
            else:
                if nums[mid] < target <= nums[right]:
                    left = mid + 1
                else:
                    right = mid - 1
        return -1

nums, target = [4,5,6,7,0,1,2], 5
a = Solution()
res = a.search(nums, target)
res
```

```python
# 34. Find First and Last Position of Element in Sorted Array
'''
- 这道题要求我们找出目标值在有序数组中的起始位置和结束位置。
- 输入是一个非递减排序的整数数组 nums 和一个目标值 target，输出是目标值的左右边界索引，如果不存在则返回 [-1, -1]。
- 最简单的方法是线性扫描，找到第一个和最后一个等于目标值的位置。
- 时间复杂度是 O(n)，空间复杂度是 O(1)，但效率不高。
- 我使用了两次二分查找的方法，分别找左边界和右边界，时间复杂度优化到 O(log n)。思路是：
    - 左边界：遇到目标值后继续向左收缩范围；
    - 右边界：遇到目标值后继续向右收缩范围；
    - 当找不到目标值时返回 [-1, -1]。
- 这道题的关键是利用二分查找分别精确定位左右边界，同时在遇到目标值时继续向一侧推进，从而准确找到范围。
'''

class Solution:
    def searchRange(self, nums: List[int], target: int):
        if not nums: return [-1, -1]
        leftbound = self.searchBound(nums, target)
        rightbound = self.searchBound(nums, target, False)
        return [leftbound, rightbound] if leftbound != -1 else [-1,-1]
    
    @staticmethod
    def searchBound(nums, target, findLeft=True):
        bound = -1
        left, right = 0, len(nums)-1
        while left <= right:
            mid = (left + right)//2
            if nums[mid] < target:
                left = mid + 1
            elif nums[mid] > target:
                right = mid - 1
            else:
                bound = mid
                if findLeft:
                    right = mid - 1
                else:
                    left = mid + 1
        return bound

nums, target = [5,7,7,8,8,10], 6
a = Solution()
res = a.searchRange(nums,target)
res
```

```python
# 35. Search Insert Position
'''
- 这道题要求我们在有序数组中查找目标值的插入位置。
- 输入是一个升序数组 nums 和一个目标值 target，输出是目标值应该插入的位置的索引。
- 最简单的方法是遍历数组找到第一个大于等于 target 的位置。
- 时间复杂度是 O(n)，空间复杂度是 O(1)。
- 我使用了标准的二分查找方法，时间复杂度优化到了 O(log n)。思路是：
    - 每次比较中间值和目标值；
    - 如果相等直接返回下标；
    - 如果中间值小于目标值，就搜索右侧；
    - 否则搜索左侧。
    - 最后返回 left 作为插入位置。
- 这道题的关键是理解二分查找的本质，最终返回 left 是因为它指向第一个大于等于目标值的位置。
'''

class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        left, right = 0, len(nums)-1
        while left <= right:
            mid = (left+right) // 2
            if nums[mid] ==  target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else: 
                right = mid - 1
        return left

nums, target = [1,3,5,6], 2
a = Solution()
res = a.searchInsert(nums, target)
res
```

```python
# 72. Edit Distance
'''
- 这道题要求我们计算将字符串 word1 转换成 word2 所需的最少操作数。
- 可以执行的操作包括插入一个字符、删除一个字符和替换一个字符。
- 输入是两个字符串 word1 和 word2，输出是它们之间的最小编辑距离。
- 最朴素的方法是使用递归暴力枚举所有可能操作组合。
- 时间复杂度是指数级，约为 O(3^(m+n))，非常低效。
- 我使用了二维动态规划的方法，将时间复杂度优化到 O(m × n)，空间复杂度也是 O(m × n)。思路如下：
    - 定义 dp[i][j] 表示 word1[0..i-1] 转换为 word2[0..j-1] 的最小操作数；
    - 初始化边界情况：dp[i][0] = i（将前 i 个字符删除），dp[0][j] = j（插入 j 个字符）；
    - 状态转移方程：
    - 若 word1[i-1] == word2[j-1]，则 dp[i][j] = dp[i-1][j-1];
    - 否则 dp[i][j] = 1 + min(插入, 删除, 替换)，即：
    - dp[i][j] = 1 + min(dp[i][j-1], dp[i-1][j], dp[i-1][j-1])。
    - 最终答案为 dp[m][n]。
- 这道题的关键是理解编辑距离的三种操作及其对应的子问题状态，动态规划可以有效避免重复子问题计算，适用于大规模输入。
'''
class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        if not word1: return len(word2)
        if not word2: return len(word1)

        # initialization
        dp = [[0]*(len(word2)+1) for _ in range(len(word1)+1)]
        for i in range(len(word1)+1):
            dp[i][0] = i
        for j in range(len(word2)+1):
            dp[0][j] = j
        
        for i in range(1, len(word1)+1):
            for j in range(1, len(word2)+1):
                if word1[i-1] == word2[j-1]:
                    dp[i][j] = dp[i-1][j-1]
                else:
                    dp[i][j] =1+min(dp[i][j-1], dp[i-1][j-1], dp[i-1][j])
        return dp[-1][-1]

word1 =  "a"; word2 = "a"
a = Solution()
res = a.minDistance(word1, word2)
res
```

```python
# 74. Search a 2D Matrix
'''
- 这道题要求我们在一个二维矩阵中查找一个整数是否存在。
- 输入是一个矩阵 matrix，其中每一行升序排列，且每行第一个元素大于前一行最后一个元素，以及一个目标值 target，输出是否找到目标值。
- 最简单的方法是暴力扫描整个矩阵。
- 时间复杂度是 O(mn)，空间复杂度是 O(1)。
- 我使用了二维转一维的二分查找，将整个矩阵视为一个一维数组，时间复杂度优化到 O(log(mn))。思路是：
    - 通过索引转换，mid // n 和 mid % n 映射到二维坐标；
    - 然后正常二分查找判断 mid 值与 target 的关系。
- 这道题的关键是掌握矩阵坐标与一维索引之间的映射关系，从而在二维结构上使用一维二分查找。
'''

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int):

        m, n = len(matrix), len(matrix[0])

        if target > matrix[m-1][n-1] or target < matrix[0][0]:
            return False
        
        left, right = 0, m*n-1
        while left <= right:
            mid = (left+right) // 2
            mid_val = matrix[mid//n][mid%n]
            if mid_val == target:
                return True
            elif mid_val > target:
                right = mid - 1
            else:
                left = mid + 1
                
        return False

matrix, target = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], 70
a = Solution()
res = a.searchMatrix(matrix, target)
res
```

```python
# 139. Word Break
'''
- 这道题要求我们判断一个字符串 s 是否可以被空格拆分成一个或多个出现在 wordDict 中的单词。
- 输入是一个字符串 s 和一个单词字典 wordDict，输出是布尔值，表示能否完成拆分。
- 最简单的方法是递归尝试所有拆分可能。
- 但这种方法时间复杂度是指数级，效率非常低。
- 我使用了动态规划，时间复杂度优化到了 O(n²)，空间复杂度是 O(n)。思路是：
    - 定义 dp[i] 表示字符串 s[:i] 是否可以被有效拆分；
    - 初始时设 dp[0] = True，表示空串是有效拆分；
    - 遍历每个位置 i，如果 dp[i] == True，说明前缀合法，就尝试从 i 向后匹配单词；
    - 如果 s[i:j] 是字典中的单词，则设置 dp[j] = True。
    - 最终返回 dp[len(s)]。
- 这道题的关键在于将原问题拆解成子问题，并且用布尔数组记录中间状态，从而避免重复计算；同时把 wordDict 转为 set 来加速查找操作。
'''
class Solution:
    def wordBreak(self, s: str, wordDict: List[str]):
        wordDict=set(wordDict)
        dp = [False] * (len(s)+1)
        dp[0] = True
        
        for i in range(len(s)):
            if dp[i]:
                for j in range(i+1, len(s)+1):
                    if s[i:j] in wordDict:
                        dp[j] = True
        return dp[len(s)]

s = "leetcode"; wordDict = ["leet","code"]
a = Solution()
res = a.wordBreak(s,wordDict)
res
```

```python
# 153. Find Minimum in Rotated Sorted Array
'''
- 这道题要求我们在一个无重复元素的旋转升序数组中找出最小值。
- 输入是一个旋转后的升序数组 nums，输出是其中的最小值。
- 最简单的方法是遍历数组找最小值。
- 时间复杂度是 O(n)，空间复杂度是 O(1)。
- 我使用了二分查找的方法，将时间复杂度优化到 O(log n)。思路是：
    - 比较中间元素和右边界；
    - 如果 nums[mid] <= nums[right]，说明最小值可能在左边，收缩 right；
    - 否则最小值在右边，收缩 left。
- 这道题的关键在于：通过与右边界比较判断最小值在哪一侧，从而用二分逻辑快速收敛。
'''
class Solution:
    def findMin(self, nums: List[int]) -> int:
        left, right = 0, len(nums)-1
        
        while left < right:    
            mid = (left+right)//2
            if nums[mid] <= nums[right]:
                right = mid
            else:
                left = mid + 1
        return nums[left]
    
nums = [1]
a = Solution()
res = a.findMin(nums)
res
```

### Binary Tree

```python
from typing import List, Dict, Tuple, Optional
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree(values: List[Optional[int]]) -> Optional[TreeNode]:
    if not values: return None
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    while queue and i < len(values):
        cur = queue.popleft()
        if values[i] is not None:
            cur.left = TreeNode(values[i])
            queue.append(cur.left)
        i += 1

        if i < len(values) and values[i] is not None:
            cur.right = TreeNode(values[i])
            queue.append(cur.right)
        i += 1
    
    return root
```

```python
# 94. Binary Tree Inorder Traversal
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        if root is None:
            return []
        return (
            self.inorderTraversal(root.left) +
            [root.val] +
            self.inorderTraversal(root.right)
        )
    def inorderTraversal_stack(self, root: Optional[TreeNode]) -> List[int]:
        res = []
        stack, start = [], root
        while stack or start:
            while start:
                stack.append(start)
                start = start.left
            cur = stack.pop()
            res.append(cur.val)
            start = cur.right
        return res

root = build_tree([1, 2, 3, None, 4, 5])
a = Solution()
res = a.inorderTraversal_stack(root)
res
```

```python
# 98. Validate Binary Search Tree
'''
- 这道题要求我们判断一棵二叉树是否是一棵合法的二叉搜索树（BST）。
- 输入是一棵二叉树的根节点 root，输出是布尔值，表示该树是否为有效的二叉搜索树。
- 二叉搜索树的定义是：每个节点的左子树只包含比当前节点小的数，右子树只包含比当前节点大的数，且每个子树也必须是 BST。
- 最直接的做法是中序遍历后判断是否为递增序列，时间复杂度是 O(n)，空间复杂度 O(n)。
- 我采用的是更优雅的递归区间法，时间复杂度是 O(n)，空间复杂度是 O(h)，其中 h 为树的高度。思路如下：
    - 给每个节点维护一个有效区间 (min, max)，判断其值是否在这个范围内；
    - 对左子树递归时，更新最大值为当前节点值；
    - 对右子树递归时，更新最小值为当前节点值；
    - 如果任意一个节点不满足约束，立即返回 False。
- 这道题的关键是：使用 上下限传递法，精确控制每个节点应该满足的范围，从而判断整棵树是否满足 BST 条件。
'''
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        if root is None: return True
        mini, maxi = float('-inf'), float('inf')
        return self.helper(root, mini, maxi)
    
    def helper(self, root, mini, maxi):
        if root is None: return True
        if not (mini< root.val < maxi): return False
        return self.helper(root.left, mini, root.val) and self.helper(root.right, root.val, maxi)
        

root = build_tree([2,1,3])
a = Solution()
res = a.isValidBST(root)
res
```

```python
# 102. Binary Tree Level Order Traversal
'''
- 这道题要求我们实现一棵二叉树的层序遍历，即从上到下逐层打印每一层的节点。
- 输入是一棵二叉树的根节点 root，输出是一个二维数组，每个子数组表示一层的节点值。
- 我使用了递归 + 深度参数模拟层序遍历的解法，时间复杂度是 O(n)，空间复杂度也是 O(n)。思路如下：
    - 定义一个辅助函数 helper(node, depth)，其中 depth 表示当前节点的层级；
    - 如果当前深度还没有对应的数组，就新增一个空数组用于存放该层节点；
    - 然后将当前节点的值加入对应层级的数组中；
    - 分别递归左子树和右子树，depth + 1.
    - 最终返回的 res 即为层序遍历结果。
- 这道题的关键是：用**递归中的深度信息替代队列**行为，实现一种递归版本的层序遍历。
'''
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        res = []
        def helper(node, depth):
            if not node: return
            if depth == len(res):
                res.append([])
            res[depth].append(node.val)
            helper(node.left, depth+1)
            helper(node.right, depth+1)
        helper(root, 0)
        return res

root = build_tree([3,9,20,None,None,15,7])
a = Solution()
res =a.levelOrder(root)
res
```

```python
# 108.Convert Sorted Array to Binary Search Tree
class Solution:
    def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
        return self.buildBST(nums, 0, len(nums)-1)

    def buildBST(self, nums, left, right):
        if left > right: return None
        if left == right: return TreeNode(nums[left])
        mid = (left+right) // 2
        root = TreeNode(nums[mid])
        root.left = self.buildBST(nums, left, mid-1)
        root.right = self.buildBST(nums, mid+1, right)
        return root

nums = [-10,-3,0,5,9]
a = Solution()
res = a.sortedArrayToBST(nums)
res
```

```python
# 113. Path Sum II
'''
可以不用recursion，用iteration配合stack来使用，做dfs/backtracking
'''
class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> List[List[int]]:
        res = []
        def backtrack(path, root,targetSum):
            if not root: return
            path.append(root.val)
            if not root.left and not root.right:
                if root.val == targetSum:
                    res.append(path[:])
            else:
                backtrack(path, root.left, targetSum-root.val)
                backtrack(path, root.right, targetSum-root.val)
            path.pop()
        backtrack([], root, targetSum)
        return res

root = build_tree([5,4,8,11,None,13,4,7,2,None,None,5,1])
targetSum = 22
a = Solution()
res = a.pathSum(root, targetSum)
res
```

```python
# 114. Flatten Binary Tree to Linked List

'''
- 检查当前节点有没有左子树。
- 有的话：
    - 暂存左、右子树。
    - 把左子树移到右边。
    - 找到左子树的最右节点。
    - 把原来的右子树接到最右节点上。
- 最后一步，往右走。
'''

class Solution:
    def flatten(self, root: Optional[TreeNode]) -> None:
        """
        Do not return anything, modify root in-place instead.
        """
        if root is None: return None
        current = root
        while current:
            if current.left:
                temp1 = current.left
                temp2 = current.right
                current.left = None
                current.right = temp1
                rightmost = temp1
                while rightmost.right:
                    rightmost = rightmost.right
                rightmost.right = temp2
            current = current.right

root = build_tree([1,2,5,3,4,None,6])
a = Solution()
res = a.flatten(root)
res
```

```python
# 124. Binary Tree Maximum Path Sum
'''
- 这道题要求我们找出二叉树中的最大路径和，路径可以从任意节点出发，到任意节点结束，但必须连续且只能向下走（父或子）。
- 输入是一棵二叉树的根节点 root，输出是所有路径中最大路径和的数值。
- 最简单的想法是枚举所有路径，但实现上复杂且效率低下。
- 我使用了递归后序遍历 + 全局变量记录最大值的方法，时间复杂度是 O(n)，空间复杂度是 O(h)，其中 n 是节点数，h 是树高。思路如下：
    - 对每个节点，递归求其左子树和右子树的最大路径贡献值（如果为负数就视为 0，因为不选更优）；
    - 然后以当前节点为“路径最高点”，路径和为：node.val + max(left, 0) + max(right, 0)；
    - 每次计算这种“经过该节点的最大路径和”，并用全局变量 self.maxi 更新最大值；
    - 返回值只需要是**“单侧路径最大贡献”**，即 max(left + node.val, right + node.val, node.val)，用于父节点决策。
- 这道题的关键在于：区分“经过当前节点的完整路径”和“从当前节点出发向上的单边路径”，一个用于更新全局最大值，一个用于递归返回。
'''
class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        self.maxi = float('-inf')
        def helper(node):
            if not node: return 0
            left = helper(node.left)
            right = helper(node.right)
            res = node.val + max(left,0) + max(right,0)
            self.maxi = max(self.maxi, res)
            return max(left+node.val, right+node.val, node.val)
        helper(root)
        return self.maxi

root = build_tree([9,6,-3,None,None,-6,2,None,None,2,None,-6,-6,-6])
a = Solution()
res = a.maxPathSum(root)
res
```

```python
# 199. Binary Tree Right Side View
'''
这道题和102层序遍历刚好相反，是同一道题。

- 道题要求我们返回从右侧观察一棵二叉树时，每一层最右边的节点值。
- 输入是一棵二叉树的根节点 root，输出是一个数组，表示每一层从右侧看到的第一个节点。
- 最直观的做法是使用 BFS 层序遍历，在每一层记录最后访问的节点。
- 时间复杂度是 O(n)，空间复杂度也是 O(n)。
- 我采用的是递归 + 右优先的深度优先遍历（DFS），时间复杂度是 O(n)，空间复杂度是 O(h)，其中 h 是树的高度。思路如下：
    - 使用辅助函数 helper(node, depth)，记录当前节点所在的层数；
    - 先访问右子树，再访问左子树，这样每一层第一个访问的节点就是右视图节点；
    - 当当前深度等于结果数组的长度时，说明是该层第一次访问节点，直接加入结果；
    - 继续递归左右子树，depth + 1。
- 这道题的关键在于：利用 DFS 右优先遍历的顺序特点，确保每层优先访问最右边的节点，并只记录第一次访问的值。
'''

class Solution:
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
        res = []
        def helper(node, depth):
            if node is None: return None
            if depth == len(res):
                res.append(node.val)
            helper(node.right, depth+1)
            helper(node.left, depth+1)
        helper(root, 0)
        return res

root = build_tree([1,2,3,None,5,None,4])
a = Solution()
res = a.rightSideView(root)
res
```

```python
# 230. Kth smallest element in BST
'''
iteration 还是好于 recursion的，第二个方法就写着玩玩。
- 这道题要求我们在一棵二叉搜索树（BST）中找到第 k 小的元素。
- 输入是一棵 BST 的根节点 root 和一个整数 k，输出是第 k 小的节点的值。
- 最简单的方法是中序遍历整个树，得到升序数组后取第 k-1 个值。
- 时间复杂度为 O(n)，空间复杂度也是 O(n)，不够高效。
- 我使用了中序遍历 + 计数器的迭代写法，在遍历过程中计数，找到第 k 个节点立即返回。
- 时间复杂度是 O(h + k)，空间复杂度是 O(h)，其中 h 是树的高度。思路如下：
    - 中序遍历 BST 会按照升序顺序访问节点；
    - 使用一个栈来模拟中序遍历过程；
    - 每访问一个节点就将计数器加一；
    - 当计数器等于 k 时，返回当前节点的值即可；
    - 这样可以在找到目标节点时提前结束遍历，避免不必要的开销。
- 这道题的关键是：利用 BST 的中序遍历特性，使得第 k 小的元素正好是遍历过程中的第 k 个访问节点。
'''
class Solution:
    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        ctm = 0
        node = root
        stack = []
        while stack or node:
            while node:
                stack.append(node)
                node = node.left
            cur = stack.pop()
            ctm += 1
            if ctm == k:
                return cur.val
            node = node.right
        
    def kthSmallest_rec(self, root: Optional[TreeNode], k: int) -> int:
        left_count = self.count_left(root.left)
        if k <= left_count:
            return self.kthSmallest_rec(root.left, k)
        elif k == left_count+1:
            return root.val
        else:
            return self.kthSmallest_rec(root.right, k-left_count-1)

    def count_left(self, root):
        if not root: return 0
        return 1 + self.count_left(root.left)+self.count_left(root.right)

root = build_tree([3,1,4,None,2])
k = 1
a = Solution()
res = a.kthSmallest_rec(root, k)
res
```

```python
# 236. Lowest Common Ancerstor of Binary Tree
'''
- 这道题要求我们找出二叉树中两个节点 p 和 q 的最近公共祖先（Lowest Common Ancestor）。
- 输入是一棵二叉树的根节点 root 和两个目标节点 p、q，输出是它们的最近公共祖先节点。
- 最直接的方法是用回溯，从叶子向上找 p 和 q，并判断它们第一次相遇的位置。
- 时间复杂度是 O(n)，空间复杂度是 O(h)，其中 n 是节点总数，h 是树的高度。
- 我使用了自顶向下的递归搜索法，思路如下：
    - 如果当前节点为 None，返回 None；
    - 如果当前节点是 p 或 q，返回当前节点；
    - 分别递归左右子树，去找 p 和 q；
    - 如果左右子树都不为空，说明 p 和 q 分别在当前节点两侧，当前节点就是最近公共祖先；
    - 否则返回非空的那一边（说明两节点在同一侧）。
- 这道题的关键在于：理解二叉树的后序遍历 + 返回值设计，并用递归自然地回溯定位到第一个“左右分叉点”。
'''
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if not root: return None
        if root == p or root == q:
            return root
        left = self.lowestCommonAncestor(root.left, p, q)
        right = self.lowestCommonAncestor(root.right, p, q)
        if left and right:
            return root
        return left or right
    
root = build_tree([3,5,1,6,2,0,8,None,None,7,4])
p, q = TreeNode(5), TreeNode(1)
a = Solution()
res = a.lowestCommonAncestor(root, p, q)
res
```

```python
# 437. Path Sum III
'''
所有记录历史的，都可以考虑prefix Sum; Prefix特别适合任意开端的sub DS

- 这道题要求我们统计二叉树中路径和等于某个值的路径个数，路径可以从任意节点出发，但必须向下（从父到子）连续连接。
- 输入是二叉树的根节点 root 和目标和 targetSum，输出是路径和为 targetSum 的路径条数。
- 最朴素的方法是以每个节点作为起点 DFS 所有路径，时间复杂度 O(n²)，在树结构稠密时效率低。
- 我使用了前缀和 + DFS 回溯的优化解法，将时间复杂度优化为 O(n)，空间复杂度也是 O(n)。思路如下：
    - 使用一个哈希表 prefix 记录从根节点到当前节点的路径中所有前缀和出现的次数；
    - 每到一个新节点，更新当前路径和 curSum = curSum + node.val；
    - 如果 curSum - targetSum 在 prefix 中出现过，说明存在从某个祖先节点到当前节点的路径和为 targetSum；
    - 将当前 curSum 加入 prefix 中，继续递归左右子树；
    - 在递归返回时，撤销当前节点的贡献，以保证前缀和状态只作用于当前路径（回溯）。
    - 最后返回路径计数。
- 这道题的关键在于：将路径问题转化为前缀和差值问题，用哈希表记录前缀状态，同时通过 DFS+回溯保证正确性和不重复计数。
'''
from collections import defaultdict

class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> int:
        # prefix
        prefix = defaultdict(int)
        prefix[0] = 1 # 从根节点出发的虚拟路径，因为根节点之前没有点，所以ancestor path sum为0
        # curSum代表现有的路径和
        def dfs(root, curSum):
            if not root: return 0
            curSum += root.val
            cnt = prefix[curSum-targetSum]
            
            prefix[curSum] += 1
            cnt += dfs(root.left, curSum)
            cnt += dfs(root.right, curSum)
            prefix[curSum] -= 1
            return cnt
        return dfs(root, 0)

root = build_tree([10,5,-3,3,2,None,11,3,-2,None,1])
targetSum = 8
a = Solution()
res = a.pathSum(root, targetSum)
res
```

```python
# 543. Diameter of Binary Tree
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        self.dia = 0
        self.dfs(root)
        return self.dia
        
    def dfs(self,root):
        if not root: return 0
        left = self.dfs(root.left)
        right = self.dfs(root.right)
        self.dia = max(self.dia, left+right)
        return max(left, right) + 1

root = build_tree([1,2,3,4,5])
a = Solution()
res = a.diameterOfBinaryTree(root)
res
```
