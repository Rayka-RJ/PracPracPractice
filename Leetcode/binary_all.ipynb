```python
# 4.Median of Two Sorted Arrays
'''
- 这道题是要求我们在 两个有序数组中找到它们的中位数。
- 输入是两个递增排序的整型数组 nums1 和 nums2，输出是它们合并后数组的中位数。
- Naive解法是将两个数组合并，然后排序，再找中位数。时间复杂度是 O((m+n)log(m+n))，因为有一个排序过程；空间复杂度是 O(m+n)，因为创建了新数组。
- 我使用了Binary Search, 时间复杂度是 O(log(min(m, n)))。具体做法是：
    - 保证 nums1 比 nums2 短，然后在 nums1 上进行二分；
    - 设定一个分割点，把两个数组一共划分成左右两部分，使得左边的最大值小于等于右边的最小值；
    - 最终根据数组总长度是奇数还是偶数，返回中位数。
- 这种方法的空间复杂度是 O(1)，因为只用了常数级变量，没有新建数组。
- 这道题的关键是转换思路，不直接合并数组，而是通过在短数组上二分寻找一个合理的划分位置。
'''

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        if len(nums1) > len(nums2):
            nums1, nums2 = nums2, nums1
        left, right = 0, len(nums1)
        while left <= right:
            i = (left + right)//2
            nums1_left_max = float('-inf') if i == 0 else nums1[i-1]
            nums1_right_min = float('inf') if i == len(nums1) else nums1[i]

            j = (len(nums1)+len(nums2)+1)//2 - i
            nums2_left_max = float('-inf') if j == 0 else nums2[j-1]
            nums2_right_min = float('inf') if j == len(nums2) else nums2[j]

            if nums1_left_max <= nums2_right_min and nums2_left_max <= nums1_right_min:
                if (len(nums1)+len(nums2)) %2 == 0:
                    return (max(nums1_left_max, nums2_left_max) + min(nums1_right_min, nums2_right_min))/2
                else:
                    return max(nums1_left_max, nums2_left_max)
            elif nums1_left_max > nums2_right_min:
                right -= 1
            else:
                left += 1

nums1 = [1,3] 
nums2 = [2]
a = Solution()
res = a.findMedianSortedArrays(nums1, nums2)
res
```

```python
# 15.3Sum
'''
- 这道题要求我们找出数组中所有不重复的三元组，使得这三个数的和为 0。
- 输入是一个整数数组 nums，输出是一个二维数组，包含所有满足条件的不重复三元组。
- 最简单的做法是三重循环枚举所有三元组，判断是否为 0。时间复杂度是 O(n³)，空间复杂度取决于去重方式，一般是 O(n)。
- 我使用了 排序 + 双指针 的方法，把时间复杂度优化到 O(n²)。具体做法是：
    - 先对数组排序；
    - 固定一个数 num，然后在其右侧用左右指针找两个数，使得三者之和为 0；
    - 为了去重，跳过重复的固定数，指针移动时也跳过重复的元素。
    - 这种方法的空间复杂度是 O(1)（不计输出结果）。
- 这道题的核心是将暴力三重循环优化为 排序 + 双指针 的双层结构，同时注意 如何跳过重复元素 保证不出现重复解。
'''

class Solution:
    def threeSum(self, nums):
        nums.sort()
        res = []
        for i, num in enumerate(nums):
            if i > 0 and num == nums[i-1]:
                continue
            two_pointer = nums[i+1:]
            j,k = 0, len(two_pointer)-1
            while j < k:
                two_sum = two_pointer[j] + two_pointer[k]
                if two_sum + num > 0:
                    k -= 1
                elif two_sum + num < 0:
                    j += 1
                else:
                    res.append([two_pointer[j], two_pointer[k], num])
                    k -= 1
                    j += 1
                    while j < k and two_pointer[j] == two_pointer[j-1]:
                        j += 1
        return res

nums = [-1,0,1,2,-1,-4]
a = Solution()
res = a.threeSum(nums)
res
```

```python
# 18. 4Sum
'''
- 这道题是要求我们找出所有不重复的四元组，使得它们的和等于目标值 target。
- 输入是一个整数数组 nums 和一个目标值 target，输出是一个二维数组，包含所有满足条件的不重复四元组。
- 最朴素的解法是使用四重循环枚举所有组合，判断是否满足条件。
- 时间复杂度是 O(n⁴)，效率很低，不适合大数据量。
- 我使用了 排序 + 双重 for 循环 + 双指针 的优化方法，把时间复杂度降到 O(n³)。具体思路是：
    - 先对数组排序；
    - 枚举前两个数 nums[i] 和 nums[j]，然后对后面的部分用双指针寻找剩下两个数；
    - 每次找到一个组合后，要跳过重复元素，避免重复解。
    - 这个方法的空间复杂度是 O(1)（除了结果集以外）。
- 这道题的关键是将四层暴力循环优化为 两层循环 + 双指针，同时通过排序和跳过重复元素来保证去重正确且高效。
'''
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        res = []
        for i in range(len(nums)):
            if i > 0 and nums[i] == nums[i-1]:continue

            for j in range(i+1, len(nums)):
                if j > i+1 and nums[j] == nums[j-1]: continue

                left, right = j+1, len(nums)-1
                while left < right:
                    sum_all = nums[i] + nums[j] + nums[left] + nums[right]
                    if sum_all > target:
                        right -= 1
                    elif sum_all < target:
                        left += 1
                    else:
                        res.append([nums[i], nums[j], nums[left], nums[right]])
                        left += 1
                        while left < right and nums[left] == nums[left-1]:
                            left += 1
                        right -= 1
                        while left < right and nums[right] == nums[right+1]:
                            right -= 1

        return res

nums, target = [1,-2,-5,-4,-3,3,3,5], -11
a = Solution()
res = a.fourSum(nums, target)
res
```

```python
# 33. Search in Rotated Sorted Array
''' 
- 这道题要求我们在一个旋转过的升序数组中查找一个目标值，如果找到返回下标，否则返回 -1。
- 输入是一个整数数组 nums（无重复，原本升序，经过一次旋转）和一个整数 target，输出是目标值的下标或 -1。
- 最简单的方法是遍历整个数组查找目标值。
- 时间复杂度是 O(n)，空间复杂度是 O(1)，但没有利用有序性质。
- 我使用了一个变种的 二分查找，时间复杂度优化到了 O(log n)。思路是：
    - 每次比较中间元素 nums[mid] 和目标值；
    - 通过判断哪一边是有序的（nums[left] <= nums[mid] 或反之），
    - 再判断 target 是否落在这个有序段中，从而决定向哪一边收缩搜索范围。
    - 这种方法空间复杂度是 O(1)，只使用了几个指针变量。
- 这道题的关键是理解旋转数组的结构特点，并且在二分查找中判断哪一段是有序的，从而排除一半搜索区间。
'''

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left, right = 0 , len(nums)-1
        while left <= right:
            mid = (left + right)//2
            if nums[mid] == target:
                return mid

            # left is sorted
            if nums[left] <= nums[mid]:
                if nums[left] <= target < nums[mid]:
                    right = mid - 1
                else:
                    left = mid + 1
            # right is sorted
            else:
                if nums[mid] < target <= nums[right]:
                    left = mid + 1
                else:
                    right = mid - 1
        return -1

nums, target = [4,5,6,7,0,1,2], 5
a = Solution()
res = a.search(nums, target)
res
```

```python
# 34. Find First and Last Position of Element in Sorted Array
'''
- 这道题要求我们找出目标值在有序数组中的起始位置和结束位置。
- 输入是一个非递减排序的整数数组 nums 和一个目标值 target，输出是目标值的左右边界索引，如果不存在则返回 [-1, -1]。
- 最简单的方法是线性扫描，找到第一个和最后一个等于目标值的位置。
- 时间复杂度是 O(n)，空间复杂度是 O(1)，但效率不高。
- 我使用了两次二分查找的方法，分别找左边界和右边界，时间复杂度优化到 O(log n)。思路是：
    - 左边界：遇到目标值后继续向左收缩范围；
    - 右边界：遇到目标值后继续向右收缩范围；
    - 当找不到目标值时返回 [-1, -1]。
- 这道题的关键是利用二分查找分别精确定位左右边界，同时在遇到目标值时继续向一侧推进，从而准确找到范围。
'''

class Solution:
    def searchRange(self, nums: List[int], target: int):
        if not nums: return [-1, -1]
        leftbound = self.searchBound(nums, target)
        rightbound = self.searchBound(nums, target, False)
        return [leftbound, rightbound] if leftbound != -1 else [-1,-1]
    
    @staticmethod
    def searchBound(nums, target, findLeft=True):
        bound = -1
        left, right = 0, len(nums)-1
        while left <= right:
            mid = (left + right)//2
            if nums[mid] < target:
                left = mid + 1
            elif nums[mid] > target:
                right = mid - 1
            else:
                bound = mid
                if findLeft:
                    right = mid - 1
                else:
                    left = mid + 1
        return bound

nums, target = [5,7,7,8,8,10], 6
a = Solution()
res = a.searchRange(nums,target)
res
```

```python
# 35. Search Insert Position
'''
- 这道题要求我们在有序数组中查找目标值的插入位置。
- 输入是一个升序数组 nums 和一个目标值 target，输出是目标值应该插入的位置的索引。
- 最简单的方法是遍历数组找到第一个大于等于 target 的位置。
- 时间复杂度是 O(n)，空间复杂度是 O(1)。
- 我使用了标准的二分查找方法，时间复杂度优化到了 O(log n)。思路是：
    - 每次比较中间值和目标值；
    - 如果相等直接返回下标；
    - 如果中间值小于目标值，就搜索右侧；
    - 否则搜索左侧。
    - 最后返回 left 作为插入位置。
- 这道题的关键是理解二分查找的本质，最终返回 left 是因为它指向第一个大于等于目标值的位置。
'''

class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        left, right = 0, len(nums)-1
        while left <= right:
            mid = (left+right) // 2
            if nums[mid] ==  target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else: 
                right = mid - 1
        return left

nums, target = [1,3,5,6], 2
a = Solution()
res = a.searchInsert(nums, target)
res
```

```python
# 74. Search a 2D Matrix
'''
- 这道题要求我们在一个二维矩阵中查找一个整数是否存在。
- 输入是一个矩阵 matrix，其中每一行升序排列，且每行第一个元素大于前一行最后一个元素，以及一个目标值 target，输出是否找到目标值。
- 最简单的方法是暴力扫描整个矩阵。
- 时间复杂度是 O(mn)，空间复杂度是 O(1)。
- 我使用了二维转一维的二分查找，将整个矩阵视为一个一维数组，时间复杂度优化到 O(log(mn))。思路是：
    - 通过索引转换，mid // n 和 mid % n 映射到二维坐标；
    - 然后正常二分查找判断 mid 值与 target 的关系。
- 这道题的关键是掌握矩阵坐标与一维索引之间的映射关系，从而在二维结构上使用一维二分查找。
'''

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int):

        m, n = len(matrix), len(matrix[0])

        if target > matrix[m-1][n-1] or target < matrix[0][0]:
            return False
        
        left, right = 0, m*n-1
        while left <= right:
            mid = (left+right) // 2
            mid_val = matrix[mid//n][mid%n]
            if mid_val == target:
                return True
            elif mid_val > target:
                right = mid - 1
            else:
                left = mid + 1
                
        return False

matrix, target = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], 70
a = Solution()
res = a.searchMatrix(matrix, target)
res
```

```python
# 153. Find Minimum in Rotated Sorted Array
'''
- 这道题要求我们在一个无重复元素的旋转升序数组中找出最小值。
- 输入是一个旋转后的升序数组 nums，输出是其中的最小值。
- 最简单的方法是遍历数组找最小值。
- 时间复杂度是 O(n)，空间复杂度是 O(1)。
- 我使用了二分查找的方法，将时间复杂度优化到 O(log n)。思路是：
    - 比较中间元素和右边界；
    - 如果 nums[mid] <= nums[right]，说明最小值可能在左边，收缩 right；
    - 否则最小值在右边，收缩 left。
- 这道题的关键在于：通过与右边界比较判断最小值在哪一侧，从而用二分逻辑快速收敛。
'''
class Solution:
    def findMin(self, nums: List[int]) -> int:
        left, right = 0, len(nums)-1
        
        while left < right:    
            mid = (left+right)//2
            if nums[mid] <= nums[right]:
                right = mid
            else:
                left = mid + 1
        return nums[left]
    
nums = [1]
a = Solution()
res = a.findMin(nums)
res
```
