## Array & Sort Algorithm

这里是一些具体sort algorithm的实现，for reference

```python
from typing import List, Dict, Tuple, Optional
```

```python
# 56. Merge Intervals
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        # step/1 sort
        intervals.sort(key=lambda x:x[0])
        # step/2 sweep
        res = []
        cur = intervals[0]
        for i in intervals:
            if i[0] <= cur[1]:
                cur[1] = max(cur[1], i[1])
            else:
                res.append(cur)
                cur = i
        res.append(cur)
        return res

interval = [[1,3],[2,6],[8,10],[15,18]]
a = Solution()
res = a.merge(interval)
res
```
