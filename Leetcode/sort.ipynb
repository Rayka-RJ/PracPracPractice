```python
# 56. Merge Intervals
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        # step/1 sort
        intervals.sort(key=lambda x:x[0])
        # step/2 sweep
        res = []
        cur = intervals[0]
        for i in intervals:
            if i[0] <= cur[1]:
                cur[1] = max(cur[1], i[1])
            else:
                res.append(cur)
                cur = i
        res.append(cur)
        return res

interval = [[1,3],[2,6],[8,10],[15,18]]
a = Solution()
res = a.merge(interval)
res
```

```python
# 75. Sort Colors
# 虽然可以直接sort，但只有三个数字，绝对值域之下，不需要sort
# 三个指针，四个区间，[0, l] - 0 , [l, l+i] - 1, [i, r] - tobeaddressed, [r:] - 2
# l 换过来可以直接动i, 但r不行，因为i走过过l，没有见过r，得重新检查该位置
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        i, l, r = 0, 0, len(nums)-1
        while i <= r:
            if nums[i] == 0:
                nums[i], nums[l] = nums[l], nums[i]
                l += 1
                i += 1
            elif nums[i] == 2:
                nums[i], nums[r] = nums[r], nums[i]
                r -= 1
            else:
                i += 1

nums = [2,0,2,1,1,0]
a = Solution()
a.sortColors(nums)
nums  
```

```python
# 189. Rotate Array
# k 可能大于len(nums),所以要先%n
# 注意id和length的区别
# 左旋右旋是取决于区间的，写一个swap函数，然后原地置换，注意互相置换的子区间范围
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        k = k % len(nums)
        # reverse function
        def reverse(nums, l, r):
            while l < r:
                nums[l], nums[r] = nums[r], nums[l]
                l += 1
                r -= 1
            return nums
        
        reverse(nums, 0, len(nums)-1)
        reverse(nums, 0, k-1)
        reverse(nums, k, len(nums)-1)   

nums = [1,2,3,4,5,6,7]
k = 3
a = Solution()
a.rotate(nums, k)
nums
```

```python
# 238. Product of Array Except Self
# prefix sum
class Solution:
    def selfthink_productExceptSelf(self, nums: List[int]) -> List[int]:
        # step/1 collect the product
        p = 1
        zero = 0
        for n in nums:
            if n != 0:
                p *= n
            else:
                zero += 1
        # step/2 replace
        for i in range(len(nums)):
            if zero == 1:
                if nums[i] != 0:
                    nums[i] = 0
                else:
                    nums[i] = p
            elif zero == 0:
                nums[i] = p//nums[i]
            else:
                nums[i] = 0
        return nums

    # 遍历list，对每个nums[i]，先乘以nums[:i]，再乘以nums[i+1:]
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        ans = [1]
        # step/1 left product as suffix
        for i in range(1, len(nums)):
            pro = ans[i-1] * nums[i-1]
            ans.append(pro)
        # step/2 right product
        pro = 1 # 替代了最端头的idx
        for i in range(len(nums)-2, -1, -1):
            pro *=nums[i+1]
            ans[i] *= pro
        return ans 
nums = [1,2,3,4]
a = Solution()
res = a.productExceptSelf(nums)
res
```

```python
# 287. Find the Duplicate Number
# 快慢指针，寻找cycle——这道题其实是链表的变体
class Solution:
    def findDuplicate(self, nums: List[int]) -> int:
        slow, fast = nums[0],nums[0]
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]]
            if slow == fast:
                break

        p1 = nums[0]
        p2 = slow
        while p1 != p2:
            p1 = nums[p1]
            p2 = nums[p2]
        return p1
nums = [3,1,3,4,2]
a = Solution()
res = a.findDuplicate(nums)
res
```
