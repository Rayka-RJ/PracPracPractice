## 20. Valid Parentheses (Easy)

```python
class Solution:
    def isValid(self, s: str) -> bool:
        # TODO: Implement your solution
        check = {'(':')', '{':'}', '[':']'}
        valid = []
        for i in s:
            if i in check:
                valid.append(i)
            else:
                if not valid:
                    return False
                j = valid.pop()
                if i != check[j]:
                    return False
        return not valid 

# Test cases
def test_isValid():
    s = Solution()
    assert s.isValid("()") is True
    assert s.isValid("()[]{}") is True
    assert s.isValid("(]") is False
    assert s.isValid("(())") is True

    print("All test cases passed!")

test_isValid()
```
