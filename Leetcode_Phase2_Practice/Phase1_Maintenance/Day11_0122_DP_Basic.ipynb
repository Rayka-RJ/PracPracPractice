```python
class Solution:
    def climbStairs(self, n: int) -> int:
        # TODO: Implement your solution
        dp = [0] * (n+1)
        dp[0], dp[1] = 1, 1
        for i in range(2, n+1):
            dp[i] = dp[i-1] + dp[i-2]
        return dp[n]

# Test cases
def test_climbStairs():
    s = Solution()
    assert s.climbStairs(2) == 2
    assert s.climbStairs(3) == 3
    assert s.climbStairs(4) == 5

    print("All test cases passed!")

test_climbStairs()
```
