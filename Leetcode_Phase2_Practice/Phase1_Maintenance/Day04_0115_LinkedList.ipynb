```python
from typing import List, Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def build_linked_list(nums: List[int]) -> Optional[ListNode]:
    """ä»Žåˆ—è¡¨æž„å»ºé“¾è¡¨"""
    dummy = ListNode()
    current = dummy
    for num in nums:
        current.next = ListNode(num)
        current = current.next
    return dummy.next

def linked_list_to_list(head: Optional[ListNode]) -> List[int]:
    """å°†é“¾è¡¨è½¬æ¢ä¸ºåˆ—è¡¨"""
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result

def print_linked_list(head: Optional[ListNode]) -> str:
    """æ‰“å°é“¾è¡¨"""
    vals = []
    while head:
        vals.append(str(head.val))
        head = head.next
    return ' -> '.join(vals) if vals else 'Empty'

# å¸¦randomæŒ‡é’ˆçš„èŠ‚ç‚¹
class RandomNode:
    def __init__(self, x: int, next: 'RandomNode' = None, random: 'RandomNode' = None):
        self.val = int(x)
        self.next = next
        self.random = random

def build_random_list(data: List[tuple]) -> Optional[RandomNode]:
    """
    æž„å»ºå¸¦randomæŒ‡é’ˆçš„é“¾è¡¨
    data: List of tuples like [(val, random_index), ...]
    Example: [(7, None), (13, 0), (11, 4), (10, 2), (1, 0)]
    """
    if not data:
        return None
    nodes = [RandomNode(val) for val, _ in data]
    for i in range(len(nodes) - 1):
        nodes[i].next = nodes[i + 1]
    for i, (_, rand_i) in enumerate(data):
        if rand_i is not None:
            nodes[i].random = nodes[rand_i]
    return nodes[0]


# Helper: build cycle list
def build_cycle_list(nums: List[int], pos: int) -> Optional[ListNode]:
    if not nums:
        return None
    head = build_linked_list(nums)
    if pos < 0:
        return head
    tail = head
    cycle_entry = None
    idx = 0
    while tail.next:
        if idx == pos:
            cycle_entry = tail
        tail = tail.next
        idx += 1
    if idx == pos:
        cycle_entry = tail
    if cycle_entry:
        tail.next = cycle_entry
    return head
```

## 206. Reverse Linked List (Easy)

```python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # TODO: Implement your solution
        prev = None
        current = head
        while current:
            temp = current.next
            current.next = prev
            prev = current
            current = temp
        return prev

# Test cases
def test_reverseList():
    s = Solution()
    head = build_linked_list([1,2,3,4,5])
    result = s.reverseList(head)
    assert linked_list_to_list(result) == [5,4,3,2,1]
    assert linked_list_to_list(s.reverseList(build_linked_list([1]))) == [1]
    assert linked_list_to_list(s.reverseList(None)) == []

    print("All test cases passed!")

test_reverseList()
```

## 21. Merge Two Sorted Lists (Easy)

```python
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # TODO: Implement your solution
        dummy = ListNode()
        current = dummy
        while list1 and list2:
            if list1.val <= list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next
            current = current.next
        while list1:
            current.next = list1
            list1 = list1.next
            current = current.next
        while list2:
            current.next = list2
            list2 = list2.next
            current = current.next
        return dummy.next

# Test cases
def test_mergeTwoLists():
    s = Solution()
    l1 = build_linked_list([1,2,4])
    l2 = build_linked_list([1,3,4])
    result = s.mergeTwoLists(l1, l2)
    assert linked_list_to_list(result) == [1,1,2,3,4,4]
    assert linked_list_to_list(s.mergeTwoLists(None, None)) == []
    assert linked_list_to_list(s.mergeTwoLists(build_linked_list([]), build_linked_list([0]))) == [0]

    print("All test cases passed!")

test_mergeTwoLists()
```
