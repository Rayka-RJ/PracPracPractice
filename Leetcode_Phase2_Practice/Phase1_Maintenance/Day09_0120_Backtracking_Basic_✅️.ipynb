````python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        # TODO: Implement your solution
        result = []
        def backtrack(path, start):
            result.append(path[:])
            for i in range(start, len(nums)):
                path.append(nums[i])
                backtrack(path, i + 1)
                path.pop()
        backtrack([], 0)
        return result

# Test cases
def test_subsets():
    s = Solution()
    result = s.subsets([1,2,3])
    assert len(result) == 8
    assert [] in result
    assert [1,2,3] in result

    print("All test cases passed!")

test_subsets()
````
