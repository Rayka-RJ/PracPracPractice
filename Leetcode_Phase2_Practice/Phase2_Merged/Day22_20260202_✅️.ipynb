````python
from typing import Optional, List
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree(values: List[Optional[int]]) -> Optional[TreeNode]:
    if not values:
        return None
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    while queue and i < len(values):
        node = queue.popleft()
        if values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    return root
````

## 230. Kth Smallest Element in a BST (Medium)

````python
# 112 path sum I
# å’Œ two sumä¸€æ ·ï¼Œç›´æ¥æ£€æŸ¥targetsum-root.valå³å¯
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        if not root: return False
        if not root.left and not root.right:
            return targetSum == root.val
        return self.hasPathSum(root.left, targetSum-root.val) or self.hasPathSum(root.right, targetSum-root.val)

root = build_tree([5,4,8,11,None,13,4,7,2,None,None, None,1])
a = Solution()
a.hasPathSum(root, targetSum=22)
````

````python
# å…¸å‹çš„dfs search
# ä¸¤ä¸ªæ­¥éª¤ï¼š 1 - æ‰¾åˆ°åˆå§‹èŠ‚ç‚¹ä½ç½®ï¼› 2 - å¯»æ‰¾åˆç†çš„path

class Solution:
    def solve_437(self, root: Optional[TreeNode], targetSum: int) -> int:
        """è¿”å›è·¯å¾„å’Œç­‰äºtargetSumçš„è·¯å¾„æ•°ç›®ã€‚"""
        if not root: return 0
        
        # å¯»æ‰¾path
        def dfs(node, currentSum):
            if not node: return 0
            count = 0
            if currentSum == node.val:
                count = 1
            count += dfs(node.left, currentSum-node.val)
            count += dfs(node.right, currentSum-node.val)
            return count

        # ä»å½“å‰èŠ‚ç‚¹å‡ºå‘çš„è·¯å¾„æ•°
        cnt = dfs(root, targetSum)
        # ä»å·¦å³å­æ ‘ä½œä¸ºæ–°èµ·ç‚¹çš„è·¯å¾„æ•° ï¼ˆæ”¹å˜åˆå§‹èŠ‚ç‚¹ä½ç½®ï¼‰
        cnt += self.solve_437(root.left, targetSum)
        cnt += self.solve_437(root.right, targetSum)
        return cnt

def test_solve_437():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_437(build_tree([10,5,-3,3,2,None,11,3,-2,None,1]), 8) == 3
    assert s.solve_437(build_tree([5,4,8,11,None,13,4,7,2,None,None,5,1]), 22) == 3
    
    # å•èŠ‚ç‚¹
    assert s.solve_437(build_tree([1]), 1) == 1
    assert s.solve_437(build_tree([1]), 0) == 0
    
    # ç©ºæ ‘
    assert s.solve_437(None, 0) == 0
    print("All test cases for 437 passed!")


test_solve_437()
````

## 543. Diameter of Binary Tree (Easy)

ç»™å®šä¸€æ£µäºŒå‰æ ‘ï¼Œä½ éœ€è¦è®¡ç®—å®ƒçš„ç›´å¾„é•¿åº¦ã€‚ä¸€æ£µäºŒå‰æ ‘çš„ç›´å¾„é•¿åº¦æ˜¯ä»»æ„ä¸¤ä¸ªç»“ç‚¹è·¯å¾„é•¿åº¦ä¸­çš„æœ€å¤§å€¼ã€‚è¿™æ¡è·¯å¾„å¯èƒ½ç©¿è¿‡ä¹Ÿå¯èƒ½ä¸ç©¿è¿‡æ ¹ç»“ç‚¹ã€‚

**æ³¨æ„ï¼š** ä¸¤ç»“ç‚¹ä¹‹é—´çš„è·¯å¾„é•¿åº¦æ˜¯ä»¥å®ƒä»¬ä¹‹é—´è¾¹çš„æ•°ç›®è¡¨ç¤ºã€‚

**Example 1:**
- Input: `root = [1,2,3,4,5]`
- Output: `3`
- è§£é‡Š: è·¯å¾„ [4,2,1,3] æˆ– [5,2,1,3] çš„é•¿åº¦ä¸º 3ã€‚

**Example 2:**
- Input: `root = [1,2]`
- Output: `1`

**Constraints:**
- æ ‘ä¸­ç»“ç‚¹æ•°åœ¨èŒƒå›´ `[1, 10^4]` å†…
- `-100 <= Node.val <= 100`

````python
# è¿™é“é¢˜å’Œæœ€å¤§dia sumæ˜¯ä¸€æ ·çš„ï¼Œä¸ºä»€ä¹ˆä¸€ä¸ªhardä¸€ä¸ªeasyï¼Œä¸ç†è§£

class Solution:
    def solve_543(self, root: Optional[TreeNode]) -> int:
        """è¿”å›äºŒå‰æ ‘çš„ç›´å¾„é•¿åº¦ã€‚"""
        self.dia = 0
        self.helper(root)
        return self.dia

    def helper(self, root):
        if not root: return 0
        left = self.helper(root.left)
        right = self.helper(root.right)
        self.dia = max(self.dia, left+right)
        return 1 + max(left, right)  
        # è¿”å›å½“å‰èŠ‚ç‚¹çš„æ·±åº¦ï¼ˆç»™çˆ¶èŠ‚ç‚¹ç”¨ï¼Œ root is Trueæ˜¯ç§¯ç´¯çˆ¶èŠ‚ç‚¹åˆ°rootçš„é‚£æ¡è¾¹ï¼‰
        # ç„¶ååœ¨leftå’Œrightä¸­äºŒé€‰ä¸€æœ€å¤§çš„ï¼Œåªè¦æ˜¯å•pathéƒ½æ¶‰åŠè¿™ç§äºŒé€‰ä¸€

def test_solve_543():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_543(build_tree([1,2,3,4,5])) == 3
    assert s.solve_543(build_tree([1,2])) == 1
    
    # å•èŠ‚ç‚¹
    assert s.solve_543(build_tree([1])) == 0
    
    # åªæœ‰å·¦å­æ ‘
    assert s.solve_543(build_tree([1,2])) == 1
    print("All test cases for 543 passed!")


test_solve_543()
````

## 108. Convert Sorted Array to Binary Search Tree (Easy)

ç»™ä½ ä¸€ä¸ªæ•´æ•°æ•°ç»„ `nums`ï¼Œå…¶ä¸­å…ƒç´ å·²ç»æŒ‰**å‡åº**æ’åˆ—ï¼Œè¯·ä½ å°†å…¶è½¬æ¢ä¸ºä¸€æ£µ**é«˜åº¦å¹³è¡¡**äºŒå‰æœç´¢æ ‘ã€‚

````python
# é«˜åº¦å¹³è¡¡äºŒå‰æœç´¢æ ‘ï¼Œè¯´æ˜inorderçš„æ—¶å€™ä¸€è·¯å¾€å‰
# root.right = self.solve_108(nums[mid+1:])   # â† åˆ‡ç‰‡åˆ›å»ºæ–°æ•°ç»„
# ä¼˜åŒ–ï¼š ç”¨left,rightå¯ä»¥ç›´æ¥éå†
class Solution:
    def solve_108_copy(self, nums: List[int]) -> Optional[TreeNode]:
        """å°†æœ‰åºæ•°ç»„è½¬æ¢ä¸ºé«˜åº¦å¹³è¡¡BSTã€‚"""
        if not nums: return None
        mid = len(nums)//2
        root = TreeNode(nums[mid])
        root.left = self.solve_108(nums[:mid])
        root.right = self.solve_108(nums[mid+1:])
        return root

    def solve_108(self, nums: List[int]) -> Optional[TreeNode]:
        """å°†æœ‰åºæ•°ç»„è½¬æ¢ä¸ºé«˜åº¦å¹³è¡¡BSTã€‚"""
        if not nums: return None
        def buildBST(left, right):
            if left > right: return None
            mid = (left+right+1)//2 #å·¦åä¸­ç‚¹æˆ–è€…å³åä¸­ç‚¹éƒ½å¯ä»¥ï¼Œç›®å‰æ˜¯ä½¿ç”¨å³åä¸­ç‚¹åŒ¹é…æµ‹è¯•
            root = TreeNode(nums[mid])
            root.left = buildBST(left, mid-1)
            root.right = buildBST(mid+1, right)
            return root
        return buildBST(0, len(nums)-1)

def test_solve_108():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    root1 = s.solve_108([-10,-3,0,5,9])
    assert root1.val == 0
    assert root1.left.val == -3
    assert root1.right.val == 9
    
    root2 = s.solve_108([1,3])
    assert root2.val in [1, 3]  # ä¸¤ç§ç­”æ¡ˆéƒ½æ­£ç¡®
    
    # å•å…ƒç´ 
    root3 = s.solve_108([1])
    assert root3.val == 1
    assert root3.left is None
    assert root3.right is None
    
    # ä¸¤ä¸ªå…ƒç´ 
    root4 = s.solve_108([1,2])
    assert root4.val in [1, 2]
    print("All test cases for 108 passed!")
    
test_solve_108()
````

````python
# 109 single-linked list ---> bst
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
def build_linked(nums:List):
    dummy = ListNode()
    curr = dummy
    for n in nums:
        curr.next = ListNode(n)
        curr = curr.next
    return dummy.next

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def sortedListToBST(self, head: Optional[ListNode]) -> Optional[TreeNode]:
        nums = []
        # æ•°å€¼æ”¶å–
        while head:
            nums.append(head.val)
            head = head.next
        
        def build(left, right):
            if left > right: return None
            mid = (left + right)//2
            node = TreeNode(nums[mid])
            node.left = build(left, mid-1)
            node.right = build(mid+1, right)
            return node
        
        return build(0, len(nums)-1)


head = build_linked([-10,-3,0,5,9])
a = Solution()
res = a.sortedListToBST(head)
res
````

## 39. Combination Sum (Medium)

ç»™ä½ ä¸€ä¸ª**æ— é‡å¤å…ƒç´ **çš„æ•´æ•°æ•°ç»„ `candidates` å’Œä¸€ä¸ªç›®æ ‡æ•´æ•° `target`ï¼Œæ‰¾å‡º `candidates` ä¸­å¯ä»¥ä½¿æ•°å­—å’Œä¸ºç›®æ ‡æ•° `target` çš„**æ‰€æœ‰ä¸åŒç»„åˆ**ï¼Œå¹¶ä»¥åˆ—è¡¨å½¢å¼è¿”å›ã€‚ä½ å¯ä»¥æŒ‰**ä»»æ„é¡ºåº**è¿”å›è¿™äº›ç»„åˆã€‚

`candidates` ä¸­çš„**åŒä¸€ä¸ªæ•°å­—å¯ä»¥æ— é™åˆ¶é‡å¤è¢«é€‰å–**ã€‚å¦‚æœè‡³å°‘ä¸€ä¸ªæ•°å­—çš„è¢«é€‰æ•°é‡ä¸åŒï¼Œåˆ™ä¸¤ç§ç»„åˆæ˜¯ä¸åŒçš„ã€‚

å¯¹äºç»™å®šçš„è¾“å…¥ï¼Œä¿è¯å’Œä¸º `target` çš„ä¸åŒç»„åˆæ•°å°‘äº `150` ä¸ªã€‚

**Example 1:**
- Input: `candidates = [2,3,6,7]`, `target = 7`
- Output: `[[2,2,3],[7]]`
- è§£é‡Š: 2 å’Œ 3 å¯ä»¥å½¢æˆä¸€ç»„å€™é€‰ï¼Œ2 + 2 + 3 = 7ã€‚æ³¨æ„ 2 å¯ä»¥ä½¿ç”¨å¤šæ¬¡ã€‚7 ä¹Ÿæ˜¯ä¸€ä¸ªå€™é€‰ï¼Œ7 = 7ã€‚ä»…æœ‰è¿™ä¸¤ç§ç»„åˆã€‚

**Example 2:**
- Input: `candidates = [2,3,5]`, `target = 8`
- Output: `[[2,2,2,2],[2,3,3],[3,5]]`

**Constraints:**
- `1 <= candidates.length <= 30`
- `2 <= candidates[i] <= 40`
- `candidates` çš„æ‰€æœ‰å…ƒç´ **äº’ä¸ç›¸åŒ**
- `1 <= target <= 40`

````python
# æ ‡å‡†backtracking
class Solution:
    def solve_39(self, candidates: List[int], target: int) -> List[List[int]]:
        """è¿”å›æ‰€æœ‰å¯ä»¥ä½¿æ•°å­—å’Œä¸ºtargetçš„ç»„åˆã€‚"""
        res = []
        def backtrack(path, start, remain):
            if remain == 0:
                res.append(path[:])
                return
            elif remain <0:
                return
            
            for i in range(start, len(candidates)):
                path.append(candidates[i])
                backtrack(path, i, remain-candidates[i])
                path.pop()
        
        backtrack([], 0, target)
        return res

def test_solve_39():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    result1 = s.solve_39([2,3,6,7], 7)
    assert len(result1) == 2
    assert [2,2,3] in result1
    assert [7] in result1
    
    result2 = s.solve_39([2,3,5], 8)
    assert len(result2) == 3
    assert [2,2,2,2] in result2
    assert [2,3,3] in result2
    assert [3,5] in result2
    
    # æ— è§£
    assert s.solve_39([2], 1) == []
    
    # å•ä¸ªå…ƒç´ 
    assert s.solve_39([1], 1) == [[1]]
    print("All test cases for 39 passed!")


test_solve_39()
````
