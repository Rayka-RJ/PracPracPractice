## 79. Word Search (Medium)

ç»™å®šä¸€ä¸ª `m x n` äºŒç»´å­—ç¬¦ç½‘æ ¼ `board` å’Œä¸€ä¸ªå­—ç¬¦ä¸²å•è¯ `word`ã€‚å¦‚æœ `word` å­˜åœ¨äºç½‘æ ¼ä¸­ï¼Œè¿”å› `True`ï¼›å¦åˆ™ï¼Œè¿”å› `False`ã€‚

å•è¯å¿…é¡»æŒ‰ç…§å­—æ¯é¡ºåºï¼Œé€šè¿‡ç›¸é‚»çš„å•å…ƒæ ¼å†…çš„å­—æ¯æ„æˆï¼Œå…¶ä¸­"ç›¸é‚»"å•å…ƒæ ¼æ˜¯é‚£äº›æ°´å¹³ç›¸é‚»æˆ–å‚ç›´ç›¸é‚»çš„å•å…ƒæ ¼ã€‚åŒä¸€ä¸ªå•å…ƒæ ¼å†…çš„å­—æ¯**ä¸å…è®¸è¢«é‡å¤ä½¿ç”¨**ã€‚

**Example 1:**
- Input: `board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]`, `word = "ABCCED"`
- Output: `True`

**Example 2:**
- Input: `board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]`, `word = "SEE"`
- Output: `True`

**Example 3:**
- Input: `board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]`, `word = "ABCB"`
- Output: `False`

**Constraints:**
- `m == board.length`
- `n = board[i].length`
- `1 <= m, n <= 6`
- `1 <= word.length <= 15`
- `board` å’Œ `word` ä»…ç”±å¤§å°å†™è‹±æ–‡å­—æ¯ç»„æˆ

````python
# è¿™é“é¢˜å’Œpermutationæœ‰ç›¸ä¼¼ä¹‹å¤„ï¼Œæ¯æ¬¡éƒ½åªmaskè‡ªå·±å³å¯ï¼Œä¸éœ€è¦ä¸€ä¸ªvisitedæ£€æŸ¥æˆ–æ§åˆ¶idxèŒƒå›´
class Solution:
    def solve_79(self, board: List[List[str]], word: str) -> bool:
        """åˆ¤æ–­å•è¯æ˜¯å¦å­˜åœ¨äºç½‘æ ¼ä¸­ã€‚"""
        # TODO: å®ç°ä½ çš„è§£æ³•ï¼ˆDFSå›æº¯ï¼‰
        def backtrack(i, j, idx):
            if idx >= len(word):
                return True
            if i<0 or i>=len(board) or j<0 or j>=len(board[0]) or board[i][j] != word[idx]:
                return False
            temp = board[i][j]
            board[i][j] = '#'
            res = backtrack(i-1,j,idx+1) or backtrack(i+1,j,idx+1) or backtrack(i, j+1, idx+1) or backtrack(i, j-1, idx+1)
            board[i][j] = temp
            return res
        for i in range(len(board)):
            for j in range(len(board[0])):
                if backtrack(i,j,0):
                    return True
        return False

def test_solve_79():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]
    assert s.solve_79(board, "ABCCED") is True
    assert s.solve_79(board, "SEE") is True
    assert s.solve_79(board, "ABCB") is False
    
    # å•å­—ç¬¦
    assert s.solve_79([["A"]], "A") is True
    assert s.solve_79([["A"]], "B") is False
    
    # å•è¡Œ
    assert s.solve_79([["A","B","C"]], "ABC") is True
    print("All test cases for 79 passed!")

test_solve_79()
````

## 17. Letter Combinations of a Phone Number (Medium)

ç»™å®šä¸€ä¸ªä»…åŒ…å«æ•°å­— `2-9` çš„å­—ç¬¦ä¸²ï¼Œè¿”å›æ‰€æœ‰å®ƒèƒ½è¡¨ç¤ºçš„å­—æ¯ç»„åˆã€‚ç­”æ¡ˆå¯ä»¥æŒ‰**ä»»æ„é¡ºåº**è¿”å›ã€‚

ç»™å‡ºæ•°å­—åˆ°å­—æ¯çš„æ˜ å°„å¦‚ä¸‹ï¼ˆä¸ç”µè¯æŒ‰é”®ç›¸åŒï¼‰ã€‚æ³¨æ„ 1 ä¸å¯¹åº”ä»»ä½•å­—æ¯ã€‚

```
2: "abc"
3: "def"
4: "ghi"
5: "jkl"
6: "mno"
7: "pqrs"
8: "tuv"
9: "wxyz"
```

**Example 1:**
- Input: `digits = "23"`
- Output: `["ad","ae","af","bd","be","bf","cd","ce","cf"]`

**Example 2:**
- Input: `digits = ""`
- Output: `[]`

**Example 3:**
- Input: `digits = "2"`
- Output: `["a","b","c"]`

**Constraints:**
- `0 <= digits.length <= 4`
- `digits[i]` æ˜¯èŒƒå›´ `['2', '9']` çš„ä¸€ä¸ªæ•°å­—

````python
# æ³¨æ„æ¯ä¸€å±‚çš„éå†é€»è¾‘
class Solution:
    def solve_17(self, digits: str) -> List[str]:
        """è¿”å›ç”µè¯å·ç çš„æ‰€æœ‰å­—æ¯ç»„åˆã€‚"""
        vocal = {'2':'abc', '3':'def', '4':'ghi', '5':'jkl', '6':'mno', '7':'pqrs', '8':'tuv', '9':'wxyz'}
        res = []
        if not digits: return []
        def backtrack(path, start):
            if len(path) == len(digits):
                res.append(''.join(path))
                return
            for i in range(start, len(digits)):
                for j in vocal[digits[i]]:
                    path.append(j)
                    backtrack(path, i+1)
                    path.pop()
        backtrack([], 0)
        return res

def test_solve_17():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    result1 = s.solve_17("23")
    assert len(result1) == 9
    assert "ad" in result1
    assert "ae" in result1
    assert "cf" in result1
    
    assert s.solve_17("") == []
    
    result2 = s.solve_17("2")
    assert result2 == ["a","b","c"]
    
    # å•ä¸ªæ•°å­—
    result3 = s.solve_17("9")
    assert len(result3) == 4
    assert "w" in result3
    assert "x" in result3
    assert "y" in result3
    assert "z" in result3
    print("All test cases for 17 passed!")


test_solve_17()
````

## 22. Generate Parentheses (Medium)

æ•°å­— `n` ä»£è¡¨ç”Ÿæˆæ‹¬å·çš„å¯¹æ•°ï¼Œè¯·ä½ è®¾è®¡ä¸€ä¸ªå‡½æ•°ï¼Œç”¨äºèƒ½å¤Ÿç”Ÿæˆæ‰€æœ‰å¯èƒ½çš„å¹¶ä¸”**æœ‰æ•ˆçš„**æ‹¬å·ç»„åˆã€‚

**Example 1:**
- Input: `n = 3`
- Output: `["((()))","(()())","(())()","()(())","()()()"]`

**Example 2:**
- Input: `n = 1`
- Output: `["()"]`

**Constraints:**
- `1 <= n <= 8`

````python
# backtracking 
# input ä¸º å½“å‰çš„path å’Œ ä¸‹ä¸€æ­¥è¦éå†çš„èŒƒå›´
# å…ˆå†™ resæ”¶å–æ¡ä»¶ å’Œ ä¸­æ–­æ¡ä»¶
# å†æŒ¨ä¸ªå¤„ç†å¯èƒ½çš„è·¯å¾„ï¼ˆæ¯”å¦‚åŠ å·¦è¾¹è¿˜æ˜¯åŠ å³è¾¹ï¼Œå‘å·¦èµ°è¿˜æ˜¯å‘å³èµ°ï¼‰

class Solution:
    def solve_22(self, n: int) -> List[str]:
        """ç”Ÿæˆnå¯¹æ‹¬å·çš„æ‰€æœ‰æœ‰æ•ˆç»„åˆã€‚"""
        res = []
        def backtrack(path, left, right):
            if len(path) == 2 * n:
                res.append(''.join(path))
                return
            if left < n:
                path.append('(')
                backtrack(path, left+1, right)
                path.pop()
            if right < left:
                path.append(')')
                backtrack(path, left, right+1)
                path.pop()
        backtrack([], 0, 0)
        return res 


def test_solve_22():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    result1 = s.solve_22(3)
    assert len(result1) == 5
    assert "((()))" in result1
    assert "(()())" in result1
    assert "()()()" in result1
    
    assert s.solve_22(1) == ["()"]
    
    # n=2
    result2 = s.solve_22(2)
    assert len(result2) == 2
    assert "(())" in result2
    assert "()()" in result2
    print("All test cases for 22 passed!")


test_solve_22()
````

## 51. N-Queens (Hard)

æŒ‰ç…§å›½é™…è±¡æ£‹çš„è§„åˆ™ï¼Œçš‡åå¯ä»¥æ”»å‡»ä¸ä¹‹å¤„åœ¨åŒä¸€è¡Œæˆ–åŒä¸€åˆ—æˆ–åŒä¸€æ–œçº¿ä¸Šçš„æ£‹å­ã€‚

**n çš‡åé—®é¢˜**ç ”ç©¶çš„æ˜¯å¦‚ä½•å°† `n` ä¸ªçš‡åæ”¾ç½®åœ¨ `nÃ—n` çš„æ£‹ç›˜ä¸Šï¼Œå¹¶ä¸”ä½¿çš‡åå½¼æ­¤ä¹‹é—´ä¸èƒ½ç›¸äº’æ”»å‡»ã€‚

ç»™ä½ ä¸€ä¸ªæ•´æ•° `n`ï¼Œè¿”å›æ‰€æœ‰ä¸åŒçš„**n çš‡åé—®é¢˜**çš„è§£å†³æ–¹æ¡ˆã€‚

æ¯ä¸€ç§è§£æ³•åŒ…å«ä¸€ä¸ªä¸åŒçš„ n çš‡åé—®é¢˜çš„æ£‹å­æ”¾ç½®æ–¹æ¡ˆï¼Œè¯¥æ–¹æ¡ˆä¸­ `'Q'` å’Œ `'.'` åˆ†åˆ«ä»£è¡¨äº†çš‡åå’Œç©ºä½ã€‚

**Example 1:**
- Input: `n = 4`
- Output: `[[".Q..","...Q","Q...","..Q."],["..Q.","Q...","...Q",".Q.."]]`
- è§£é‡Š: å¦‚ä¸Šå›¾æ‰€ç¤ºï¼Œ4 çš‡åé—®é¢˜å­˜åœ¨ä¸¤ä¸ªä¸åŒçš„è§£æ³•ã€‚

**Example 2:**
- Input: `n = 1`
- Output: `[["Q"]]`

**Constraints:**
- `1 <= n <= 9`

````python
class Solution:
    def solve_51(self, n: int) -> List[List[str]]:
        """è¿”å›nçš‡åé—®é¢˜çš„æ‰€æœ‰è§£ã€‚"""
        # TODO: å®ç°ä½ çš„è§£æ³•ï¼ˆå›æº¯ï¼‰
        res = []
        board = [['.'] * n for _ in range(n)]
        cols = set()
        dia1 = set()
        dia2 = set()
        def backtrack(k, board, cols, dia1, dia2):
            if k == n:
                res.append([''.join(b) for b in board])
                return
            for j in range(n):
                if j in cols or k-j in dia1 or k+j in dia2:
                    continue
                board[k][j] = 'Q'
                cols.add(j)
                dia1.add(k-j)
                dia2.add(k+j)
                backtrack(k+1, board, cols, dia1, dia2)
                board[k][j] = '.'
                cols.remove(j)
                dia1.remove(k-j)
                dia2.remove(k+j)
        backtrack(0, board, cols, dia1, dia2)
        return res


def test_solve_51():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    result1 = s.solve_51(4)
    assert len(result1) == 2
    # éªŒè¯æ¯ä¸ªè§£éƒ½æ˜¯nè¡Œnåˆ—
    for sol in result1:
        assert len(sol) == 4
        for row in sol:
            assert len(row) == 4
            assert row.count('Q') == 1
    
    assert s.solve_51(1) == [["Q"]]
    
    # n=2å’Œn=3æ— è§£
    assert s.solve_51(2) == []
    assert s.solve_51(3) == []
    print("All test cases for 51 passed!")


test_solve_51()
````

## 131. Palindrome Partitioning (Medium)

ç»™ä½ ä¸€ä¸ªå­—ç¬¦ä¸² `s`ï¼Œè¯·ä½ å°† `s` åˆ†å‰²æˆä¸€äº›å­ä¸²ï¼Œä½¿æ¯ä¸ªå­ä¸²éƒ½æ˜¯**å›æ–‡ä¸²**ã€‚è¿”å› `s` æ‰€æœ‰å¯èƒ½çš„åˆ†å‰²æ–¹æ¡ˆã€‚

**Example 1:**
- Input: `s = "aab"`
- Output: `[["a","a","b"],["aa","b"]]`

**Example 2:**
- Input: `s = "a"`
- Output: `[["a"]]`

**Constraints:**
- `1 <= s.length <= 16`
- `s` ä»…ç”±å°å†™è‹±æ–‡å­—æ¯ç»„æˆ

````python
class Solution:
    def solve_131(self, s: str) -> List[List[str]]:
        """è¿”å›å­—ç¬¦ä¸²çš„æ‰€æœ‰å›æ–‡åˆ†å‰²æ–¹æ¡ˆã€‚"""
        res = []
        def backtrack(path, start):
            if start >= len(s):
                res.append(path[:])
                return
            for i in range(start, len(s)):
                left, right = start, i
                is_mirror = True
                while left <= right:
                    if s[left] != s[right]:
                        is_mirror = False
                        break
                    left += 1
                    right -= 1
                if is_mirror:
                    path.append(s[start:i+1])
                    backtrack(path, i+1)
                    path.pop()
        backtrack([], 0)
        return res

def test_solve_131():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    result1 = s.solve_131("aab")
    assert len(result1) == 2
    assert ["a","a","b"] in result1
    assert ["aa","b"] in result1
    
    assert s.solve_131("a") == [["a"]]
    
    # å•ä¸ªå­—ç¬¦
    result2 = s.solve_131("aa")
    assert ["a","a"] in result2
    assert ["aa"] in result2
    
    # å›æ–‡å­—ç¬¦ä¸²
    result3 = s.solve_131("aba")
    assert ["a","b","a"] in result3
    assert ["aba"] in result3
    print("All test cases for 131 passed!")


test_solve_131()
````
