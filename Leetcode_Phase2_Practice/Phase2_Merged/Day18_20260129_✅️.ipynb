## 84. Largest Rectangle in Histogram (Hard)

ç»™å®šä¸€ä¸ªè¡¨ç¤ºç›´æ–¹å›¾çš„éè´Ÿæ•´æ•°æ•°ç»„ `heights`ï¼Œå…¶ä¸­æ¯ä¸ªå…ƒç´ è¡¨ç¤ºå®½åº¦ä¸º 1 çš„æŸ±å­çš„é«˜åº¦ï¼Œæ±‚èƒ½å¤Ÿåœ¨è¯¥ç›´æ–¹å›¾ä¸­å½¢æˆçš„æœ€å¤§çŸ©å½¢é¢ç§¯ã€‚

**Example 1:**
- Input: `heights = [2,1,5,6,2,3]`
- Output: `10`
- Explanation: æœ€å¤§çŸ©å½¢å¯ä»¥è¦†ç›–é«˜åº¦ä¸º 5 å’Œ 6 çš„ä¸¤æ ¹æŸ±å­ï¼Œæ€»å®½åº¦ä¸º 2ï¼Œé«˜åº¦ä¸º 5ï¼Œé¢ç§¯ä¸º 10ã€‚

**Example 2:**
- Input: `heights = [2,4]`
- Output: `4`

**Constraints:**
- `1 <= len(heights) <= 10^5`
- `0 <= heights[i] <= 10^4`

```python
'''
æ—¶é—´å¤æ‚åº¦ O(n)ï¼Œç©ºé—´ O(n)ã€‚
'''

class Solution:
    def solve_84(self, heights: List[int]) -> int:
        """è¿”å›ç›´æ–¹å›¾ä¸­æœ€å¤§çŸ©å½¢çš„é¢ç§¯ã€‚

        :param heights: æ¯ä¸ªå…ƒç´ æ˜¯æŸ±å­çš„é«˜åº¦ï¼Œå®½åº¦å‡ä¸º 1
        :return: èƒ½å¤Ÿå½¢æˆçš„æœ€å¤§çŸ©å½¢é¢ç§¯
        """
        # TODO: å®ç°ä½ çš„è§£æ³•
        largest = 0
        stack = []
        for idx in range(len(heights)):
            while stack and heights[stack[-1]] > heights[idx]:
                cur = stack.pop()
                left = stack[-1] if stack else -1
                area = heights[cur] * (idx-left-1)
                largest = max(largest, area)
            stack.append(idx)

        while stack:
            cur = stack.pop()
            left = stack[-1] if stack else -1
            largest = max(largest, heights[cur]*(len(heights)-left-1))
        return largest


def test_solve_84():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_84([2,1,5,6,2,3]) == 10
    assert s.solve_84([2,4]) == 4
    # è¾¹ç•Œç”¨ä¾‹
    assert s.solve_84([1]) == 1
    assert s.solve_84([]) == 0
    assert s.solve_84([0,0,0]) == 0
    print("All test cases for 84 passed!")


test_solve_84()
```

## 155. Min Stack (Easy)

è®¾è®¡ä¸€ä¸ªæ”¯æŒåœ¨å¸¸æ•°æ—¶é—´å†…å®Œæˆ `push`ã€`pop`ã€`top`ï¼Œä»¥åŠåœ¨å¸¸æ•°æ—¶é—´å†…æ£€ç´¢æœ€å°å…ƒç´  `getMin` çš„æ ˆæ•°æ®ç»“æ„ã€‚

ä½ éœ€è¦å®ç°ä¸€ä¸ª `MinStack` ç±»ï¼š
- `MinStack()` åˆå§‹åŒ–æ ˆå¯¹è±¡ï¼›
- `void push(int val)` å°†å…ƒç´  `val` å‹å…¥æ ˆä¸­ï¼›
- `void pop()` åˆ é™¤æ ˆé¡¶å…ƒç´ ï¼›
- `int top()` è·å–æ ˆé¡¶å…ƒç´ ï¼›
- `int getMin()` æ£€ç´¢æ ˆä¸­çš„æœ€å°å…ƒç´ ã€‚

**Example:**
- Input:
  - `MinStack minStack = new MinStack();`
  - `minStack.push(-2);`
  - `minStack.push(0);`
  - `minStack.push(-3);`
  - `minStack.getMin();` // è¿”å› -3
  - `minStack.pop();`
  - `minStack.top();`   // è¿”å› 0
  - `minStack.getMin();` // è¿”å› -2

**Constraints:**
- `-2^31 <= val <= 2^31 - 1`
- æ“ä½œæ€»æ•°ä¸è¶…è¿‡ \(10^4\) æ¬¡

```python
# æ·±åˆ»ç†è§£ä¸‹è¿™é‡Œ self.miniçš„å«ä¹‰ï¼šè¡¨ç¤ºåˆ°è¿™ä¸ªæ ˆå…ƒç´ ä¸ºæ­¢ï¼Œ[0:top]èŒƒå›´å†…çš„æœ€å°å€¼
# å‡è®¾stackæ˜¯[1,2,3,4,5], self.miniå°±ä¼šæ˜¯[1,1,1,1,1]ï¼Œæ‰€ä»¥topè¢«ç§»é™¤åï¼Œè¿”å›çš„æ˜¯[1,2,3,4]é‡Œé¢çš„æœ€å°å€¼
# æ‰€ä»¥å§‹ç»ˆå­˜å‚¨å½“ä¸‹æ•°ç»„çš„æœ€å°å€¼

class MinStack:
    def __init__(self):
        """åˆå§‹åŒ–æ•°æ®ç»“æ„ã€‚"""
        # TODO: å®ç°ä½ çš„æ•°æ®ç»“æ„
        self.stack = []
        self.mini = []

    def push(self, val: int) -> None:
        """å°†å…ƒç´  val å‹å…¥æ ˆä¸­ã€‚"""
        # TODO: å®ç°ä½ çš„æ–¹æ³•
        self.stack.append(val)
        if self.mini:
            self.mini.append(min(val, self.mini[-1]))
        else:
            self.mini.append(val)

    def pop(self) -> None:
        """åˆ é™¤æ ˆé¡¶çš„å…ƒç´ ã€‚"""
        # TODO: å®ç°ä½ çš„æ–¹æ³•
        if self.stack:
            self.stack.pop()
            self.mini.pop()
        else:
            return

    def top(self) -> int:
        """è·å–æ ˆé¡¶å…ƒç´ ã€‚"""
        # TODO: å®ç°ä½ çš„æ–¹æ³•
        return self.stack[-1] 

    def getMin(self) -> int:
        """æ£€ç´¢æ ˆä¸­çš„æœ€å°å…ƒç´ ã€‚"""
        # TODO: å®ç°ä½ çš„æ–¹æ³•
        return self.mini[-1]


def test_MinStack():
    ms = MinStack()
    # åŸºæœ¬æ“ä½œæµç¨‹ï¼ˆæ¥æºäºé¢˜ç›®ç¤ºä¾‹ï¼‰
    ms.push(-2)
    ms.push(0)
    ms.push(-3)
    assert ms.getMin() == -3
    ms.pop()
    assert ms.top() == 0
    assert ms.getMin() == -2

    # å…¶ä»–è¾¹ç•Œæƒ…å†µ
    ms2 = MinStack()
    ms2.push(1)
    ms2.push(2)
    ms2.push(-1)
    assert ms2.getMin() == -1
    ms2.pop()
    assert ms2.getMin() == 1

    print("All test cases for 155 passed!")


test_MinStack()
```

## 215. Kth Largest Element in an Array (Medium)

ç»™å®šä¸€ä¸ªæ•´æ•°æ•°ç»„ `nums` å’Œä¸€ä¸ªæ•´æ•° `k`ï¼Œè¯·ä½ è¿”å›æ•°ç»„ä¸­ç¬¬ `k` ä¸ªæœ€å¤§çš„å…ƒç´ ï¼ˆæŒ‰æ’åºåçš„é¡ºåºï¼Œç¬¬ 1 ä¸ªæœ€å¤§å…ƒç´ æ˜¯æœ€å¤§å€¼ï¼‰ã€‚

æ³¨æ„ï¼šä½ éœ€è¦æ‰¾çš„æ˜¯æ’åºåç¬¬ `k` ä¸ªæœ€å¤§çš„å…ƒç´ ï¼Œè€Œä¸æ˜¯ç¬¬ `k` ä¸ªä¸åŒçš„å…ƒç´ ã€‚

**Example 1:**
- Input: `nums = [3,2,1,5,6,4]`, `k = 2`
- Output: `5`

**Example 2:**
- Input: `nums = [3,2,3,1,2,4,5,5,6]`, `k = 4`
- Output: `4`

**Constraints:**
- `1 <= k <= len(nums) <= 10^5`
- `-10^4 <= nums[i] <= 10^4`

```python
'''
Can you solve it without sorting?
'''
import heapq

class Solution:
    def solve_215(self, nums: List[int], k: int) -> int:
        """è¿”å›æ•°ç»„ä¸­ç¬¬ k å¤§çš„å…ƒç´ ã€‚

        :param nums: æ•´æ•°æ•°ç»„
        :param k: ç¬¬ k å¤§ï¼ˆ1 è¡¨ç¤ºæœ€å¤§ï¼‰
        :return: ç¬¬ k å¤§å…ƒç´ çš„å€¼
        """
        # TODO: å®ç°ä½ çš„è§£æ³•
        keep = []
        for num in nums:
            heapq.heappush(keep, num)
            if len(keep) > k:
                heapq.heappop(keep)
        return keep[0]

def test_solve_215():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_215([3,2,1,5,6,4], 2) == 5
    assert s.solve_215([3,2,3,1,2,4,5,5,6], 4) == 4
    # æœ‰é‡å¤å…ƒç´ 
    assert s.solve_215([1,1,1,1], 1) == 1
    # è´Ÿæ•°ä¸æ­£æ•°æ··åˆ
    assert s.solve_215([-1, -2, 0, 3], 1) == 3
    assert s.solve_215([-1, -2, 0, 3], 4) == -2
    print("All test cases for 215 passed!")


test_solve_215()
```

## 347. Top K Frequent Elements (Medium)

ç»™å®šä¸€ä¸ªæ•´æ•°æ•°ç»„ `nums` å’Œä¸€ä¸ªæ•´æ•° `k`ï¼Œè¯·ä½ è¿”å›å…¶ä¸­å‡ºç°é¢‘ç‡å‰ `k` é«˜çš„å…ƒç´ ã€‚è¿”å›é¡ºåºå¯ä»¥ä»»æ„ã€‚

**Example 1:**
- Input: `nums = [1,1,1,2,2,3]`, `k = 2`
- Output: `[1,2]`

**Example 2:**
- Input: `nums = [1]`, `k = 1`
- Output: `[1]`

**Constraints:**
- `1 <= len(nums) <= 10^5`
- `-10^4 <= nums[i] <= 10^4`
- `1 <= k <=` ä¸åŒå…ƒç´ çš„ä¸ªæ•°

```python
'''
Your algorithm's time complexity must be better than O(n log n), where n is the array's size.

heapå°±å¾ˆé€‚åˆè¿™æ ·çš„é¢˜ï¼Œè¦åœ¨ä¸èƒ½sortçš„æƒ…å†µä¸‹æ‰¾æœ€å¤§æœ€å°å€¼ï¼Œä¹Ÿé€‚åˆé™åˆ¶é•¿åº¦ä¸ºKçš„é¢˜ã€‚å †çš„æ—¶é—´æ˜¯O(logK)
'''
import heapq

class Solution:
    def solve_347(self, nums: List[int], k: int) -> List[int]:
        """è¿”å›å‡ºç°é¢‘ç‡å‰ k é«˜çš„å…ƒç´ åˆ—è¡¨ï¼Œé¡ºåºå¯ä»¥ä»»æ„ã€‚

        :param nums: æ•´æ•°æ•°ç»„
        :param k: éœ€è¦è¿”å›çš„é«˜é¢‘å…ƒç´ ä¸ªæ•°
        :return: é•¿åº¦ä¸º k çš„å…ƒç´ åˆ—è¡¨
        """
        # TODO: å®ç°ä½ çš„è§£æ³•
        check = dict()
        for num in nums:
            check[num] = check.get(num, 0) + 1
        
        k_pair = []
        for key,v in check.items():
            heapq.heappush(k_pair, (v,key))
            if len(k_pair) > k:
                heapq.heappop(k_pair)
        return sorted([key for (v,key) in k_pair])

def test_solve_347():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert sorted(s.solve_347([1,1,1,2,2,3], 2)) == [1,2]
    assert s.solve_347([1], 1) == [1]
    # æ‰€æœ‰å…ƒç´ é¢‘ç‡ç›¸åŒ
    assert sorted(s.solve_347([1,2,3,4], 2)) in ([[1,2],[1,3],[1,4],[2,3],[2,4],[3,4]])
    # åŒ…å«è´Ÿæ•°
    res = sorted(s.solve_347([-1,-1,-2,-2,-3], 2))
    assert res in ([ -2, -1], [ -1, -2])
    print("All test cases for 347 passed!")


test_solve_347()
```

## 394. Decode String (Medium)

ç»™å®šä¸€ä¸ªç»è¿‡ç¼–ç çš„å­—ç¬¦ä¸² `s`ï¼Œè§„åˆ™æ˜¯ï¼š`k[encoded_string]` è¡¨ç¤ºå…¶ä¸­çš„ `encoded_string` æ­£å¥½é‡å¤ `k` æ¬¡ã€‚æ³¨æ„ `k` æ˜¯ä¸€ä¸ªæ­£æ•´æ•°ï¼Œä¸”å…è®¸åµŒå¥—ç¼–ç ã€‚

è¯·è§£ç å­—ç¬¦ä¸²å¹¶è¿”å›è§£ç åçš„ç»“æœå­—ç¬¦ä¸²ã€‚

**Example 1:**
- Input: `s = "3[a]2[bc]"`
- Output: `"aaabcbc"`

**Example 2:**
- Input: `s = "3[a2[c]]"`
- Output: `"accaccacc"`

**Example 3:**
- Input: `s = "2[abc]3[cd]ef"`
- Output: `"abcabccdcdcdef"`

**Constraints:**
- `1 <= len(s) <= 3 * 10^5`
- `s` ä»…ç”±å°å†™è‹±æ–‡å­—æ¯ã€æ•°å­—å’Œæ–¹æ‹¬å· `'[', ']'` ç»„æˆ
- è¾“å…¥ä¿è¯æ€»æ˜¯åˆæ³•çš„ç¼–ç å­—ç¬¦ä¸²

```python
# é€’å½’å¤ªå¤æ‚äº†ï¼Œè¿˜æ˜¯ç”¨stackæ¯”è¾ƒå¥½
# ä¹‹å‰æƒ³recursionæ˜¯è§‰å¾—K[abc]è¿™æ ·å¾ˆé€‚åˆï¼Œä½†å¥½åƒè¿™é“é¢˜å¹¶ä¸ä¸¥æ ¼æŒ‰ç…§æ ¼å¼æ¥ï¼Œå¦‚æœK=1ï¼Œå®ƒç›´æ¥ä¸å†™äº†ï¼Œå°±'ab2[a]'
# ä»è¿™ä¸ªè§’åº¦æƒ³ï¼Œå†™recursionå°±å˜å¤æ‚äº†

class Solution:
    def solve_394(self, s: str) -> str:
        """è§£ç å½¢å¦‚ k[encoded_string] çš„ç¼–ç å­—ç¬¦ä¸²ã€‚"""
        # TODO: å®ç°ä½ çš„è§£æ³•
        stack = []
        cnt = 0
        alpha = ''
        for ch in s:
            if ch.isdigit():
                cnt = cnt * 10 + int(ch)
            elif ch.isalpha():
                alpha += ch
            elif ch == '[':
                stack.append((alpha, cnt)) # update current alpha & cnt, current alphaæ˜¯å·²ç»å¤„ç†å®Œäº†çš„
                cnt = 0
                alpha = ''
            elif ch == ']':
                prev, num = stack.pop()
                alpha = prev + alpha * num
        return alpha

# åœ¨è¿™ä¸ªå‡½æ•°é‡Œï¼Œæ¯æ¬¡å…¥æ ˆçš„éƒ½æ˜¯å‰é¢å¤„ç†å®Œçš„å†åŠ ä¸Šä¸‹ä¸€ä¸ªè¦*çš„å­—æ¯ä¸²
# æ ˆé‡Œå­˜å‚¨çš„æ˜¯ prev_string + ç°åœ¨è¦ä¹˜çš„é‡ ï¼Œæ„Ÿè§‰è®¡ç®—æœºå°±æ˜¯ç›¸æ¯”å®Œæ•´pair '()' æ›´å–œæ¬¢å­˜ ')('è¿™ç§ä¸Šä¸‹æ¥å£

def test_solve_394():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_394("3[a]2[bc]") == "aaabcbc"
    assert s.solve_394("3[a2[c]]") == "accaccacc"
    assert s.solve_394("2[abc]3[cd]ef") == "abcabccdcdcdef"
    # å•å±‚ä¸ç©ºä¸²æƒ…å†µ
    assert s.solve_394("10[a]") == "aaaaaaaaaa"
    assert s.solve_394("abc") == "abc"
    print("All test cases for 394 passed!")


'''
å¯¹äºè¾“å…¥ "3[a2[c]]"ï¼Œæ¯è½®å­—ç¬¦å¤„ç†æ—¶ stack å˜åŒ–å¦‚ä¸‹ï¼š

1. '3' â†’ æ— åŠ¨ä½œï¼ˆcnt = 3ï¼‰
2. '[' â†’ stack=[('', 3)]
3. 'a' â†’ alpha = 'a'
4. '2' â†’ æ— åŠ¨ä½œï¼ˆcnt = 2ï¼‰
5. '[' â†’ stack=[('', 3), ('a', 2)]; alpha='', cnt=0
6. 'c' â†’ alpha='c'
7. ']' â†’ å¼¹å‡º('a', 2)ï¼Œalpha = 'a' + 'c'*2 = 'acc'ï¼Œstack=[('', 3)]
8. ']' â†’ å¼¹å‡º('', 3)ï¼Œalpha = '' + 'acc'*3 = 'accaccacc'ï¼Œstack=[]

æœ€ç»ˆ stack ä¸ºç©ºï¼Œç»“æœä¸º "accaccacc"ã€‚
'''

test_solve_394()
```
