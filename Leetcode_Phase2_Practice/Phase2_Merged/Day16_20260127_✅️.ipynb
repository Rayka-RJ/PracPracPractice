```python
from typing import List, Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def build_linked_list(nums: List[int]) -> Optional[ListNode]:
    dummy = ListNode()
    current = dummy
    for num in nums:
        current.next = ListNode(num)
        current = current.next
    return dummy.next

def linked_list_to_list(head: Optional[ListNode]) -> List[int]:
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result

def build_cycle_list(nums: List[int], pos: int) -> Optional[ListNode]:
    if not nums:
        return None
    head = build_linked_list(nums)
    if pos < 0:
        return head
    tail = head
    cycle_entry = None
    idx = 0
    while tail.next:
        if idx == pos:
            cycle_entry = tail
        tail = tail.next
        idx += 1
    if idx == pos:
        cycle_entry = tail
    if cycle_entry:
        tail.next = cycle_entry
    return head

class RandomNode:
    def __init__(self, x: int, next: 'RandomNode' = None, random: 'RandomNode' = None):
        self.val = int(x)
        self.next = next
        self.random = random

def build_random_list(data: List[tuple]) -> Optional[RandomNode]:
    if not data:
        return None
    nodes = [RandomNode(val) for val, _ in data]
    for i in range(len(nodes) - 1):
        nodes[i].next = nodes[i + 1]
    for i, (_, rand_i) in enumerate(data):
        if rand_i is not None:
            nodes[i].random = nodes[rand_i]
    return nodes[0]
```

## 42. Trapping Rain Water (Hard)

Given n non-negative integers representing an elevation map where the width of each bar is 1, 
compute how much water it can trap after raining.

**Example:**
- Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
- Output: 6


```python
class Solution:
    def trap(self, height: List[int]) -> int:
        # TODO: Implement your solution
        left, right = 0, len(height)-1
        left_max, right_max = 0, 0
        area = 0
        while left < right:
            if height[left] <= height[right]:
                if left_max <= height[left]:
                    left_max = height[left]
                else:
                    area += left_max - height[left]
                left += 1
            else:
                if right_max <= height[right]:
                    right_max = height[right]
                else:
                    area += right_max - height[right]
                right -= 1
        return area

# Test cases
def test_trap():
    s = Solution()
    assert s.trap([0,1,0,2,1,0,1,3,2,1,2,1]) == 6
    assert s.trap([4,2,0,3,2,5]) == 9
    assert s.trap([1,2,3]) == 0
    print("All test cases passed!")

test_trap()
```

## 19. Remove Nth Node From End of List (Medium)

Given the head of a linked list, remove the nth node from the end of the list and return its head.

**Example:**
- Input: head = [1,2,3,4,5], n = 2
- Output: [1,2,3,5]


```python
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        # TODO: Implement your solution
        dummy = ListNode(0, head)
        slow = fast = dummy
        # å…ˆèµ°
        for _ in range(n):
            fast = fast.next
        # move
        while fast.next:
            slow = slow.next
            fast = fast.next
        # remove
        temp = slow
        slow = slow.next
        temp.next = slow.next
        return dummy.next

# Test cases
def test_removeNthFromEnd():
    s = Solution()
    head = build_linked_list([1,2,3,4,5])
    result = s.removeNthFromEnd(head, 2)
    assert linked_list_to_list(result) == [1,2,3,5]
    head = build_linked_list([1])
    assert linked_list_to_list(s.removeNthFromEnd(head, 1)) == []
    head = build_linked_list([1,2])
    assert linked_list_to_list(s.removeNthFromEnd(head, 1)) == [1]
    print("All test cases passed!")

test_removeNthFromEnd()
```
