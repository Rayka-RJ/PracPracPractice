## 77. Combinations (Medium)

ç»™å®šä¸¤ä¸ªæ•´æ•° `n` å’Œ `k`ï¼Œè¿”å›èŒƒå›´ `[1, n]` ä¸­æ‰€æœ‰å¯èƒ½çš„ `k` ä¸ªæ•°çš„ç»„åˆã€‚

ä½ å¯ä»¥æŒ‰ **ä»»ä½•é¡ºåº** è¿”å›ç­”æ¡ˆã€‚

**Example 1:**
- Input: `n = 4`, `k = 2`
- Output: `[[1,2],[1,3],[1,4],[2,3],[2,4],[3,4]]`
- è§£é‡Š: 4 ä¸ªæ•°ä¸­é€‰ 2 ä¸ªæ•°çš„æ‰€æœ‰ç»„åˆã€‚

**Example 2:**
- Input: `n = 1`, `k = 1`
- Output: `[[1]]`

**Constraints:**
- `1 <= n <= 20`
- `1 <= k <= n`

```python
class Solution:
    def solve_77(self, n: int, k: int) -> List[List[int]]:
        """è¿”å›èŒƒå›´ [1, n] ä¸­æ‰€æœ‰å¯èƒ½çš„ k ä¸ªæ•°çš„ç»„åˆã€‚"""
        res = []
        def backtrack(path, start):
            if len(path) >= k:
                res.append(path[:])
                return
            for i in range(start, n+1):
                path.append(i)
                backtrack(path, i+1)
                path.pop()
        backtrack([], 1)    
        return res

def test_solve_77():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    result = s.solve_77(4, 2)
    assert len(result) == 6
    assert [1, 2] in result
    assert [1, 3] in result
    assert [1, 4] in result
    assert [2, 3] in result
    assert [2, 4] in result
    assert [3, 4] in result
    
    assert s.solve_77(1, 1) == [[1]]
    
    # è¾¹ç•Œæƒ…å†µ
    assert s.solve_77(5, 1) == [[1], [2], [3], [4], [5]]
    assert s.solve_77(3, 3) == [[1, 2, 3]]
    print("All test cases for 77 passed!")


test_solve_77()
```

## 300. Longest Increasing Subsequence (Medium)

ç»™ä½ ä¸€ä¸ªæ•´æ•°æ•°ç»„ `nums`ï¼Œæ‰¾åˆ°å…¶ä¸­æœ€é•¿ä¸¥æ ¼é€’å¢å­åºåˆ—çš„é•¿åº¦ã€‚

**å­åºåˆ—** æ˜¯ç”±æ•°ç»„æ´¾ç”Ÿè€Œæ¥çš„åºåˆ—ï¼Œåˆ é™¤ï¼ˆæˆ–ä¸åˆ é™¤ï¼‰æ•°ç»„ä¸­çš„å…ƒç´ è€Œä¸æ”¹å˜å…¶ä½™å…ƒç´ çš„é¡ºåºã€‚ä¾‹å¦‚ï¼Œ`[3,6,2,7]` æ˜¯æ•°ç»„ `[0,3,1,6,2,2,7]` çš„å­åºåˆ—ã€‚

**Example 1:**
- Input: `nums = [10,9,2,5,3,7,101,18]`
- Output: `4`
- è§£é‡Š: æœ€é•¿é€’å¢å­åºåˆ—æ˜¯ `[2,3,7,101]`ï¼Œå› æ­¤é•¿åº¦ä¸º `4`ã€‚

**Example 2:**
- Input: `nums = [0,1,0,3,2,3]`
- Output: `4`

**Example 3:**
- Input: `nums = [7,7,7,7,7,7,7]`
- Output: `1`

**Constraints:**
- `1 <= nums.length <= 2500`
- `-10^4 <= nums[i] <= 10^4`

```python
# DPé¢˜ç›®ï¼šåˆå§‹åŒ–ã€å¦‚ä½•å–æœ€ç»ˆç­”æ¡ˆã€å¤æ‚åº¦åˆ†æ

class Solution:
    def solve_3000(self, nums: List[int]) -> int:
        """è¿”å›æœ€é•¿ä¸¥æ ¼é€’å¢å­åºåˆ—çš„é•¿åº¦ã€‚"""
        dp = [1]*len(nums)
        for i in range(1, len(nums)):
            for j in range(i):
                if nums[j]<nums[i]:
                    dp[i] = max(dp[i], dp[j]+1)
        return max(dp)

    def solve_300(self, nums:List[int]) -> int:
        '''
        ç»´æŠ¤ä¸€ä¸ªæœ€é•¿çš„é€’å¢å­æ•°ç»„
        '''
        tails = []

        # ç›®æ ‡ï¼šæ‰¾åˆ°ç¬¬ä¸€ä¸ª >= target çš„ä½ç½®
        def binarysearch(tails):
            left, right = 0, len(tails)-1 
            while left <= right:
                mid = (left + right)//2
                if tails[mid] >= n:
                    right = mid - 1
                else:
                    left = mid + 1
            return left

        for n in nums:
            if not tails: tails.append(n)
            pos = binarysearch(tails)
            if pos >= len(tails):
                tails.append(n)
            else:
                tails[pos] = n
        return len(tails)
                

def test_solve_300():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_300([10,9,2,5,3,7,101,18]) == 4
    assert s.solve_300([0,1,0,3,2,3]) == 4
    assert s.solve_300([7,7,7,7,7,7,7]) == 1
    
    # å•å…ƒç´ 
    assert s.solve_300([1]) == 1
    
    # å®Œå…¨é€’å¢
    assert s.solve_300([1,2,3,4,5]) == 5
    
    # å®Œå…¨é€’å‡
    assert s.solve_300([5,4,3,2,1]) == 1
    print("All test cases for 300 passed!")


test_solve_300()
```

## 72. Edit Distance (Medium)

ç»™ä½ ä¸¤ä¸ªå•è¯ `word1` å’Œ `word2`ï¼Œè¯·è¿”å›å°† `word1` è½¬æ¢æˆ `word2` æ‰€ä½¿ç”¨çš„æœ€å°‘æ“ä½œæ•°ã€‚

ä½ å¯ä»¥å¯¹ä¸€ä¸ªå•è¯è¿›è¡Œå¦‚ä¸‹ä¸‰ç§æ“ä½œï¼š
- æ’å…¥ä¸€ä¸ªå­—ç¬¦
- åˆ é™¤ä¸€ä¸ªå­—ç¬¦
- æ›¿æ¢ä¸€ä¸ªå­—ç¬¦

**Example 1:**
- Input: `word1 = "horse"`, `word2 = "ros"`
- Output: `3`
- è§£é‡Š: 
  - `horse -> rorse` (å°† 'h' æ›¿æ¢ä¸º 'r')
  - `rorse -> rose` (åˆ é™¤ 'r')
  - `rose -> ros` (åˆ é™¤ 'e')

**Example 2:**
- Input: `word1 = "intention"`, `word2 = "execution"`
- Output: `5`
- è§£é‡Š: 
  - `intention -> inention` (åˆ é™¤ 't')
  - `inention -> enention` (å°† 'i' æ›¿æ¢ä¸º 'e')
  - `enention -> exention` (å°† 'n' æ›¿æ¢ä¸º 'x')
  - `exention -> exection` (å°† 'n' æ›¿æ¢ä¸º 'c')
  - `exection -> execution` (æ’å…¥ 'u')

**Constraints:**
- `0 <= word1.length, word2.length <= 500`
- `word1` å’Œ `word2` ç”±å°å†™è‹±æ–‡å­—æ¯ç»„æˆ

```python
# dp[i][j] = å°† word1 çš„å‰ i ä¸ªå­—ç¬¦è½¬æ¢æˆ word2 çš„å‰ j ä¸ªå­—ç¬¦æ‰€éœ€çš„æœ€å°æ“ä½œæ•°
# æ›¿æ¢ï¼šword1[:i] å˜æˆ word2[:j]ï¼Œæ„å‘³ç€word1[:i-1]å·²ç»å˜æˆäº†word2[:j-1]
# åˆ é™¤: word1[:i] å˜æˆ word1[:i-1], æ„å‘³ç€ word1[:i-1]å·²ç»æ˜¯word2[:j]ï¼Œåº”è¯¥æ˜¯dp[i-1][j] + 1
# å¢åŠ ï¼šword1[:i] å˜æˆ word1[:i+1]ï¼Œæ„å‘³ç€ word1[:i+1]ä¸ºword2[:j], åº”è¯¥æ˜¯dp[i][j-1] + 1 ä»£è¡¨dp[i+1][j]æ»¡è¶³è¦æ±‚

class Solution:
    def solve_72(self, word1: str, word2: str) -> int:
        """è¿”å›å°† word1 è½¬æ¢æˆ word2 æ‰€ä½¿ç”¨çš„æœ€å°‘æ“ä½œæ•°ã€‚"""
        # =========å­¦ä¹ è¯¥è¡¨ç¤ºæ–¹æ³•=========
        # å¤šåŠ ä¸€è¡Œç©ºä¸²
        dp = [[0] * (len(word2)+1) for _ in range(len(word1)+1)]
        for i in range(1, len(word2)+1):
            dp[0][i] = i
        for i in range(1, len(word1)+1):
            dp[i][0] = i
        # ===============================
        for i in range(1, len(word1)+1):
            for j in range(1, len(word2)+1):
                if word1[i-1] == word2[j-1]:
                    dp[i][j] = dp[i-1][j-1] #å­—ç¬¦ç›¸åŒï¼Œæ— éœ€æ“ä½œ
                else:
                    dp[i][j] = min(
                        dp[i-1][j-1] + 1,
                        dp[i-1][j] + 1,
                        dp[i][j-1] + 1
                    )
        return dp[-1][-1]
        
def test_solve_72():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_72("horse", "ros") == 3
    assert s.solve_72("intention", "execution") == 5
    
    # ç›¸åŒå­—ç¬¦ä¸²
    assert s.solve_72("abc", "abc") == 0
    
    # ä¸€ä¸ªä¸ºç©º
    assert s.solve_72("", "abc") == 3
    assert s.solve_72("abc", "") == 3
    
    # å•å­—ç¬¦
    assert s.solve_72("a", "b") == 1
    assert s.solve_72("a", "a") == 0
    print("All test cases for 72 passed!")


test_solve_72()
```

## 53. Maximum Subarray (Easy)

ç»™ä½ ä¸€ä¸ªæ•´æ•°æ•°ç»„ `nums`ï¼Œè¯·ä½ æ‰¾å‡ºä¸€ä¸ªå…·æœ‰æœ€å¤§å’Œçš„è¿ç»­å­æ•°ç»„ï¼ˆå­æ•°ç»„æœ€å°‘åŒ…å«ä¸€ä¸ªå…ƒç´ ï¼‰ï¼Œè¿”å›å…¶æœ€å¤§å’Œã€‚

**å­æ•°ç»„** æ˜¯æ•°ç»„ä¸­çš„ä¸€ä¸ªè¿ç»­éƒ¨åˆ†ã€‚

**Example 1:**
- Input: `nums = [-2,1,-3,4,-1,2,1,-5,4]`
- Output: `6`
- è§£é‡Š: è¿ç»­å­æ•°ç»„ `[4,-1,2,1]` çš„å’Œæœ€å¤§ï¼Œä¸º `6`ã€‚

**Example 2:**
- Input: `nums = [1]`
- Output: `1`

**Example 3:**
- Input: `nums = [5,4,-1,7,8]`
- Output: `23`

**Constraints:**
- `1 <= nums.length <= 10^5`
- `-10^4 <= nums[i] <= 10^4`

```python
class Solution:
    def solve_53(self, nums: List[int]) -> int:
        """è¿”å›å…·æœ‰æœ€å¤§å’Œçš„è¿ç»­å­æ•°ç»„çš„æœ€å¤§å’Œã€‚"""
        dp = [nums[0]] * len(nums)
        for i in range(1, len(nums)):
            dp[i] = max(dp[i-1]+nums[i], nums[i])
        return max(dp)
'''
nums[i]
dp[i] -- within subarray, start a new array, grab the max
'''
def test_solve_53():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_53([-2,1,-3,4,-1,2,1,-5,4]) == 6
    assert s.solve_53([1]) == 1
    assert s.solve_53([5,4,-1,7,8]) == 23
    
    # å…¨è´Ÿæ•°
    assert s.solve_53([-2,-1,-3]) == -1
    
    # å…¨æ­£æ•°
    assert s.solve_53([1,2,3,4]) == 10
    
    # å•å…ƒç´ 
    assert s.solve_53([-1]) == -1
    assert s.solve_53([0]) == 0
    print("All test cases for 53 passed!")


test_solve_53()
```

## 62. Unique Paths (Medium)

ä¸€ä¸ªæœºå™¨äººä½äºä¸€ä¸ª `m x n` ç½‘æ ¼çš„å·¦ä¸Šè§’ï¼ˆèµ·å§‹ç‚¹åœ¨ä¸‹å›¾ä¸­æ ‡è®°ä¸º "Start"ï¼‰ã€‚

æœºå™¨äººæ¯æ¬¡åªèƒ½å‘ä¸‹æˆ–è€…å‘å³ç§»åŠ¨ä¸€æ­¥ã€‚æœºå™¨äººè¯•å›¾è¾¾åˆ°ç½‘æ ¼çš„å³ä¸‹è§’ï¼ˆåœ¨ä¸‹å›¾ä¸­æ ‡è®°ä¸º "Finish"ï¼‰ã€‚

é—®æ€»å…±æœ‰å¤šå°‘æ¡ä¸åŒçš„è·¯å¾„ï¼Ÿ

**Example 1:**
- Input: `m = 3`, `n = 7`
- Output: `28`

**Example 2:**
- Input: `m = 3`, `n = 2`
- Output: `3`
- è§£é‡Š: ä»å·¦ä¸Šè§’å¼€å§‹ï¼Œæ€»å…±æœ‰ 3 æ¡è·¯å¾„å¯ä»¥åˆ°è¾¾å³ä¸‹è§’ã€‚
  1. å‘å³ -> å‘ä¸‹ -> å‘ä¸‹
  2. å‘ä¸‹ -> å‘ä¸‹ -> å‘å³
  3. å‘ä¸‹ -> å‘å³ -> å‘ä¸‹

**Example 3:**
- Input: `m = 7`, `n = 3`
- Output: `28`

**Example 4:**
- Input: `m = 3`, `n = 3`
- Output: `6`

**Constraints:**
- `1 <= m, n <= 100`
- é¢˜ç›®æ•°æ®ä¿è¯ç­”æ¡ˆå°äºç­‰äº `2 * 10^9`

```python
class Solution:
    def solve_62(self, m: int, n: int) -> int:
        """è¿”å›ä»å·¦ä¸Šè§’åˆ°å³ä¸‹è§’çš„ä¸åŒè·¯å¾„æ•°ã€‚"""
        dp = [[1]*n for _ in range(m)]
        for i in range(1, m):
            for j in range(1, n):
                dp[i][j] = dp[i-1][j] + dp[i][j-1]
        return dp[-1][-1]

def test_solve_62():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_62(3, 7) == 28
    assert s.solve_62(3, 2) == 3
    assert s.solve_62(7, 3) == 28
    assert s.solve_62(3, 3) == 6
    
    # è¾¹ç•Œæƒ…å†µ
    assert s.solve_62(1, 1) == 1
    assert s.solve_62(1, 10) == 1
    assert s.solve_62(10, 1) == 1
    print("All test cases for 62 passed!")


test_solve_62()
```
