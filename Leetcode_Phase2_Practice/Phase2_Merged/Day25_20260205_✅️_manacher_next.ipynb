## 322. Coin Change (Medium)

ç»™ä½ ä¸€ä¸ªæ•´æ•°æ•°ç»„ `coins`ï¼Œè¡¨ç¤ºä¸åŒé¢é¢çš„ç¡¬å¸ï¼›ä»¥åŠä¸€ä¸ªæ•´æ•° `amount`ï¼Œè¡¨ç¤ºæ€»é‡‘é¢ã€‚

è®¡ç®—å¹¶è¿”å›å¯ä»¥å‡‘æˆæ€»é‡‘é¢æ‰€éœ€çš„**æœ€å°‘çš„ç¡¬å¸ä¸ªæ•°**ã€‚å¦‚æœæ²¡æœ‰ä»»ä½•ä¸€ç§ç¡¬å¸ç»„åˆèƒ½ç»„æˆæ€»é‡‘é¢ï¼Œè¿”å› `-1`ã€‚

ä½ å¯ä»¥è®¤ä¸ºæ¯ç§ç¡¬å¸çš„æ•°é‡æ˜¯æ— é™çš„ã€‚

**Example 1:**
- Input: `coins = [1,2,5]`, `amount = 11`
- Output: `3`
- è§£é‡Š: `11 = 5 + 5 + 1`

**Example 2:**
- Input: `coins = [2]`, `amount = 3`
- Output: `-1`

**Example 3:**
- Input: `coins = [1]`, `amount = 0`
- Output: `0`

**Constraints:**
- `1 <= coins.length <= 12`
- `1 <= coins[i] <= 2^31 - 1`
- `0 <= amount <= 10^4`

````python
# åˆæœ€å°åˆæ•°é‡ä¸å›ºå®šï¼Œè‚¯å®šæ˜¯dp
# coins ä¸è¦ sortedï¼Œgreedyè¡Œä¸é€šï¼Œä¸æ˜¯è¶Šå¤§è¶Šå¥½
# DP å’Œè´ªå¿ƒæ˜¯ä¸¤ç§ä¸åŒçš„æ€ç»´æ–¹å¼ï¼š
# è´ªå¿ƒ (Greedy)ï¼šæ¯æ­¥éƒ½åšå±€éƒ¨æœ€ä¼˜é€‰æ‹©ï¼Œå¸Œæœ›å¾—åˆ°å…¨å±€æœ€ä¼˜ï¼ˆä½†ä¸ä¸€å®šæ­£ç¡®ï¼‰
# DPï¼šæšä¸¾æ‰€æœ‰å¯èƒ½ï¼Œé€šè¿‡å­é—®é¢˜æœ€ä¼˜è§£æ¨å¯¼å…¨å±€æœ€ä¼˜

# å¯¹äºé‡‘é¢ iï¼Œä½ éœ€è¦å°è¯•ä½¿ç”¨æ¯ä¸€ç§ç¡¬å¸ä½œä¸º"æœ€åä¸€æš"
# è¿™ä¸ªæœ‰ç‚¹ç±»ä¼¼ä¹‹å‰é‚£ä¸ªï¼šç”¨idxæ¥ä»£è¡¨valueï¼Œå› ä¸ºidxåˆšå¥½å°±æ¯”valueå¤šäº†ä¸€ä¸ª0ï¼Œæ˜¯å¯ä»¥ç”¨äºå­˜å‚¨valueçš„
# ç”¨idxå˜åŒ–æ¥ä»£æ›¿valueçš„å˜åŒ–

class Solution:
    def solve_322(self, coins: List[int], amount: int) -> int:
        """è¿”å›å‡‘æˆæ€»é‡‘é¢æ‰€éœ€çš„æœ€å°‘ç¡¬å¸ä¸ªæ•°ã€‚"""
        # TODO: å®ç°ä½ çš„è§£æ³•
        dp = [float('inf')] * (amount+1)
        dp[0] = 0
        for i in range(1, amount+1):
            for coin in coins:
                if i >= coin:
                    dp[i] = min(dp[i], dp[i-coin]+1)
        return dp[-1] if dp[-1] != float('inf') else -1

def test_solve_322():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_322([1,2,5], 11) == 3
    assert s.solve_322([2], 3) == -1
    assert s.solve_322([1], 0) == 0
    
    # æ— æ³•å‡‘æˆ
    assert s.solve_322([2,5], 1) == -1
    
    # éœ€è¦å¤šä¸ªç›¸åŒç¡¬å¸
    assert s.solve_322([1,3,4], 6) == 2  # 3+3
    print("All test cases for 322 passed!")


test_solve_322()
````

## 152. Maximum Product Subarray (Medium)

ç»™ä½ ä¸€ä¸ªæ•´æ•°æ•°ç»„ `nums`ï¼Œè¯·ä½ æ‰¾å‡ºæ•°ç»„ä¸­ä¹˜ç§¯æœ€å¤§çš„éç©ºè¿ç»­å­æ•°ç»„ï¼ˆè¯¥å­æ•°ç»„ä¸­è‡³å°‘åŒ…å«ä¸€ä¸ªæ•°å­—ï¼‰ï¼Œå¹¶è¿”å›è¯¥å­æ•°ç»„æ‰€å¯¹åº”çš„ä¹˜ç§¯ã€‚

æµ‹è¯•ç”¨ä¾‹çš„ç­”æ¡ˆæ˜¯ä¸€ä¸ª**32ä½æ•´æ•°**ã€‚

**Example 1:**
- Input: `nums = [2,3,-2,4]`
- Output: `6`
- è§£é‡Š: å­æ•°ç»„ `[2,3]` æœ‰æœ€å¤§ä¹˜ç§¯ `6`ã€‚

**Example 2:**
- Input: `nums = [-2,0,-1]`
- Output: `0`
- è§£é‡Š: ç»“æœä¸èƒ½ä¸º `2`ï¼Œå› ä¸º `[-2,-1]` ä¸æ˜¯è¿ç»­å­æ•°ç»„ã€‚

**Constraints:**
- `1 <= nums.length <= 2 * 10^4`
- `-10 <= nums[i] <= 10`
- `nums` çš„ä»»ä½•å‰ç¼€æˆ–åç¼€çš„ä¹˜ç§¯éƒ½**ä¿è¯**åœ¨ 32 ä½æ•´æ•°èŒƒå›´å†…

````python
# æ ‡å‡†åŠ¨æ€è§„åˆ’ï¼Œéç©ºå­æ•°ç»„ï¼Œæ±‚**æœ€å¤§**ä¹˜ç§¯ï¼Œæ— æ•°ä¸ªæ²¡å¤´æ²¡å°¾çš„å­é—®é¢˜
# è¿™æ¶‰åŠä¸€ä¸ªè´Ÿè´Ÿå¾—æ­£çš„é—®é¢˜ï¼Œæ¯”å¦‚ nums[i:j] æ˜¯æ­£çš„ï¼Œnums[i-1:j]æ˜¯è´Ÿï¼Œnums[i:j+1]æ˜¯è´Ÿï¼Œä½†nums[i-1:j+1]å°±æ˜¯æ­£çš„äº†
# dp[i] çš„åˆ†æ”¯æ˜¯ æ˜¯å¦ä¹˜ä»¥è‡ªå·± æˆ– å¦èµ·ä¸€æ®µå­æ•°ç»„
# å…ˆå°è¯•äºŒç»´åŠ¨æ€è§„åˆ’ï¼Œæ—¢ç„¶æ˜¯åˆ‡ç‰‡ï¼Œé‚£éƒ½æœ‰å¯èƒ½æ˜¯äºŒç»´dpï¼Œdp[i][j]è¡¨ç¤ºnums[i:j]çš„ä¹˜ç§¯
'''
æ ¸å¿ƒï¼šæ‰¾åˆ°ä»¥nums[j]ç»“å°¾çš„æœ€å¤§å’Œæœ€å°ä¹˜ç§¯ï¼Œç”¨æ¥é¿å…è´Ÿè´Ÿå¾—æ­£

å¯¹æ¯”ï¼ŒåŠ æ³•çš„å•è°ƒæ€§ï¼Œä¹˜æ³•çš„éå•è°ƒæ€§
DP è®¾è®¡åŸåˆ™ï¼šçŠ¶æ€çš„ç»´åº¦å–å†³äº"æœªæ¥å¯èƒ½éœ€è¦çš„å†å²ä¿¡æ¯"
'''

class Solution:
    def solve_152(self, nums: List[int]) -> int:
        """è¿”å›æ•°ç»„ä¸­ä¹˜ç§¯æœ€å¤§çš„è¿ç»­å­æ•°ç»„çš„ä¹˜ç§¯ã€‚"""
        # TODO: å®ç°ä½ çš„è§£æ³•
        dp_max = [float('-inf')] * len(nums) 
        dp_min = [float('inf')] * len(nums)
        dp_max[0], dp_min[0] = nums[0], nums[0]
        for i in range(1, len(nums)):
            dp_max[i] = max(nums[i], nums[i]*dp_max[i-1], nums[i]*dp_min[i-1])
            dp_min[i] = min(nums[i], nums[i]*dp_max[i-1], nums[i]*dp_min[i-1])
        return max(dp_max)

def test_solve_152():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_152([2,3,-2,4]) == 6
    assert s.solve_152([-2,0,-1]) == 0
    
    # å…¨è´Ÿæ•°
    assert s.solve_152([-2,-3,-1]) == 6
    
    # å•å…ƒç´ 
    assert s.solve_152([-2]) == -2
    assert s.solve_152([0]) == 0
    
    # åŒ…å«0
    assert s.solve_152([-2,1,0]) == 1
    print("All test cases for 152 passed!")


test_solve_152()
````

## 207. Course Schedule (Medium)

ä½ è¿™ä¸ªå­¦æœŸå¿…é¡»é€‰ä¿® `numCourses` é—¨è¯¾ç¨‹ï¼Œè®°ä¸º `0` åˆ° `numCourses - 1`ã€‚

åœ¨é€‰ä¿®æŸäº›è¯¾ç¨‹ä¹‹å‰éœ€è¦ä¸€äº›å…ˆä¿®è¯¾ç¨‹ã€‚å…ˆä¿®è¯¾ç¨‹æŒ‰æ•°ç»„ `prerequisites` ç»™å‡ºï¼Œå…¶ä¸­ `prerequisites[i] = [ai, bi]`ï¼Œè¡¨ç¤ºå¦‚æœè¦å­¦ä¹ è¯¾ç¨‹ `ai` åˆ™**å¿…é¡»**å…ˆå­¦ä¹ è¯¾ç¨‹ `bi`ã€‚

- ä¾‹å¦‚ï¼Œå…ˆä¿®è¯¾ç¨‹å¯¹ `[0, 1]` è¡¨ç¤ºï¼šæƒ³è¦å­¦ä¹ è¯¾ç¨‹ `0`ï¼Œä½ éœ€è¦å…ˆå®Œæˆè¯¾ç¨‹ `1`ã€‚

è¯·ä½ åˆ¤æ–­æ˜¯å¦å¯èƒ½å®Œæˆæ‰€æœ‰è¯¾ç¨‹çš„å­¦ä¹ ï¼Ÿå¦‚æœå¯ä»¥ï¼Œè¿”å› `true`ï¼›å¦åˆ™ï¼Œè¿”å› `false`ã€‚

**Example 1:**
- Input: `numCourses = 2`, `prerequisites = [[1,0]]`
- Output: `true`
- è§£é‡Š: æ€»å…±æœ‰ 2 é—¨è¯¾ç¨‹ã€‚å­¦ä¹ è¯¾ç¨‹ 1 ä¹‹å‰ï¼Œä½ éœ€è¦å®Œæˆè¯¾ç¨‹ 0ã€‚è¿™æ˜¯å¯èƒ½çš„ã€‚

**Example 2:**
- Input: `numCourses = 2`, `prerequisites = [[1,0],[0,1]]`
- Output: `false`
- è§£é‡Š: æ€»å…±æœ‰ 2 é—¨è¯¾ç¨‹ã€‚å­¦ä¹ è¯¾ç¨‹ 1 ä¹‹å‰ï¼Œä½ éœ€è¦å…ˆå®Œæˆè¯¾ç¨‹ 0ï¼›å¹¶ä¸”å­¦ä¹ è¯¾ç¨‹ 0 ä¹‹å‰ï¼Œä½ è¿˜åº”å…ˆå®Œæˆè¯¾ç¨‹ 1ã€‚è¿™æ˜¯ä¸å¯èƒ½çš„ã€‚

**Constraints:**
- `1 <= numCourses <= 2000`
- `0 <= prerequisites.length <= 5000`
- `prerequisites[i].length == 2`
- `0 <= ai, bi < numCourses`
- `prerequisites[i]` ä¸­çš„æ‰€æœ‰è¯¾ç¨‹å¯¹**äº’ä¸ç›¸åŒ**

````python
# dfs + backtracking
# è¿™ä¸ªæ¨¡å¼ï¼ˆä¸¤ä¸ª set + DFSï¼‰åœ¨æ‹“æ‰‘æ’åºã€ä»»åŠ¡è°ƒåº¦ã€ä¾èµ–åˆ†æä¸­éƒ½å¾ˆå¸¸è§ï¼Œè®°ä½è¿™ä¸ªæ¨¡æ¿ã€‚

class Solution:
    def solve_207(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        """åˆ¤æ–­æ˜¯å¦å¯èƒ½å®Œæˆæ‰€æœ‰è¯¾ç¨‹çš„å­¦ä¹ ã€‚"""
        # TODO: å®ç°ä½ çš„è§£æ³•
        # step 1 - æ„å»ºé‚»æ¥è¡¨ï¼ˆprerequisite graphï¼‰
        graph = {i:[] for i in range(numCourses)}
        for course, prereq in prerequisites:
            graph[course].append(prereq)

        visited = set()
        visiting = set()

        # step 2 - å¯¹æ¯ä¸ªèŠ‚ç‚¹è°ƒç”¨dfs
        # check cycle
        def is_cycle(node):
            if node in visited: return False
            if node in visiting: return True

            visiting.add(node)
            for neighbor in graph[node]:
                if is_cycle(neighbor):
                    return True
            visiting.remove(node)
            visited.add(node)
            return False

        for n in range(numCourses):
            if is_cycle(n):
                return False
        return True

def test_solve_207():
    s = Solution()
    # ç¤ºä¾‹ç”¨ä¾‹
    assert s.solve_207(2, [[1,0]]) == True
    assert s.solve_207(2, [[1,0],[0,1]]) == False
    
    # æ— å…ˆä¿®è¯¾ç¨‹
    assert s.solve_207(1, []) == True
    assert s.solve_207(3, []) == True
    
    # å¤æ‚ä¾èµ–
    assert s.solve_207(4, [[1,0],[2,1],[3,2]]) == True
    assert s.solve_207(3, [[0,1],[1,2],[2,0]]) == False
    print("All test cases for 207 passed!")


test_solve_207()
````
