## 128. Longest Consecutive Sequence (Medium)

Given an unsorted array of integers nums, return the length of the longest consecutive elements sequence.
You must write an algorithm that runs in O(n) time.

**Example:**
- Input: nums = [100,4,200,1,3,2]
- Output: 4 (The longest consecutive sequence is [1, 2, 3, 4])


```python
# O(n)è¯æ˜å¯ä»¥éå†å¸¸æ•°æ¬¡çš„list
# å…³é”®ç‚¹åœ¨äºï¼Œå¦‚æœnumæ˜¯èµ·ç‚¹ï¼Œå°±å¯¹numè¿›è¡Œç´¯åŠ ï¼Œçœ‹èƒ½åŠ åˆ°ä»€ä¹ˆç¨‹åº¦

class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        # TODO: Implement your solution
        if len(nums) < 1: return 0
        check = set(nums)
        leng = 1
        for n in check: #å°ç»†èŠ‚ï¼Œè¿™é‡Œåªç”¨unique element checkå°±å¯ä»¥ï¼Œä¸éœ€è¦éå†æ¯ä¸ªlistå…ƒç´ 
            if n-1 not in check:
                j = n
                while j in check:
                    j += 1
                leng = max(leng, j-n)
        return leng

# Test cases
def test_longestConsecutive():
    s = Solution()
    assert s.longestConsecutive([100,4,200,1,3,2]) == 4
    assert s.longestConsecutive([0,3,7,2,5,8,4,6,0,1]) == 9
    assert s.longestConsecutive([]) == 0
    print("All test cases passed!")

test_longestConsecutive()
```

## 560. Subarray Sum Equals K (Medium)

Given an array of integers nums and an integer k, 
return the total number of subarrays whose sum equals to k.

**Example:**
- Input: nums = [1,1,1], k = 2
- Output: 2


```python

class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        # TODO: Implement your solution
        check = {0:1} #ä¸€å®šè¦ç»Ÿè®¡ä¸€ä¸ªå‰ç¼€å’Œä¸º0ï¼Œä½œä¸ºinitialization
        presum = 0 # æµ‹è¯•å½“å‰å’Œ-kåœ¨ä¹‹å‰çš„presumé‡Œé¢ï¼Œåˆ™å¯ä»¥åŠ å…¥ä¹‹å‰presumçš„ä¸ªæ•°
        cnt = 0
        for i in range(len(nums)):
            presum += nums[i]
            if presum-k in check:
                cnt += check.get(presum-k,0)
            check[presum] = check.get(presum,0) + 1
        return cnt

# Test cases 
def test_subarraySum():
    s = Solution()
    assert s.subarraySum([1,1,1], 2) == 2
    assert s.subarraySum([1,2,3], 3) == 2
    assert s.subarraySum([1], 0) == 0
    print("All test cases passed!")

test_subarraySum()
```

## 76. Minimum Window Substring (Hard)

Given two strings s and t, return the minimum window substring of s 
such that every character in t (including duplicates) is included in the window.

**Example:**
- Input: s = "ADOBECODEBANC", t = "ABC"
- Output: "BANC"


```python
from collections import Counter

class Solution:
    def minWindow(self, s: str, t: str) -> str:
        # TODO: Implement your solution
        left, right = 0, 0
        check = Counter(t)
        window = {}
        mini = float('inf')
        res = (0,0)
        formed, required = 0, len(check)
        # leftï¼Œrightæ˜¯æ ‡å‡†æ»‘åŠ¨çª—å£æŒ‡é’ˆ
        # check, required éƒ½æ˜¯ç±»æ¯”ç”¨çš„target
        # windowsï¼Œformed éƒ½æ˜¯å®æ—¶æ›´æ–°çš„è¦ç´ 
        # miniï¼Œres æ˜¯ç”¨äºå¯»æ‰¾æœ€å°å€¼æœ€ç»ˆreturnçš„å…ƒç´ 

        # right extension
        while right < len(s):
            window[s[right]] = window.get(s[right], 0) + 1
            if s[right] in check and check[s[right]] == window[s[right]]: # æ£€æŸ¥è¿˜å·®å¤šå°‘ä¸ªï¼Œå…ˆç¡®å®šåœ¨é‡Œé¢ï¼Œå†ç¡®å®šæ•°é‡æ˜¯å¦ç›¸åŒ
                formed += 1 # æ¯å¢åŠ ä¸€ä¸ªä»£è¡¨è¯¥å­—æ¯å·²ç»å½»åº•æå®šäº†
            right += 1
            # left shrink
            while formed == required: # æ‰€æœ‰æ»¡è¶³æ¡ä»¶çš„å­—æ¯éƒ½é‡åˆ°äº†
                if right - left < mini: # æ›´æ–°æœ€å°å€¼
                    mini = right-left 
                    res = (left,right)
                window[s[left]] -= 1 # å·¦æ”¶
                if s[left] in check and window[s[left]] < check[s[left]]:
                    formed -= 1 # è·³å‡ºshrink loop
                left += 1
        
        return s[res[0]:res[1]] if mini != float('inf') else ''

# Test cases
def test_minWindow():
    s = Solution()
    assert s.minWindow("ADOBECODEBANC", "ABC") == "BANC"
    assert s.minWindow("a", "a") == "a"
    assert s.minWindow("a", "aa") == ""
    print("All test cases passed!")

test_minWindow()
```

## 239. Sliding Window Maximum (Hard)

You are given an array of integers nums, there is a sliding window of size k 
which is moving from the very left of the array to the very right.
Return the max sliding window.

**Example:**
- Input: nums = [1,3,-1,-3,5,3,6,7], k = 3
- Output: [3,3,5,5,6,7]


```python
from typing import List
from collections import deque

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        # TODO: Implement your solution
        window = deque()
        res = []
        for i in range(len(nums)):
            # å¼¹å‡ºå·¦ä¾§ä¸ç¬¦åˆè¦æ±‚çš„
            while window and window[0] <= i-k:
                window.popleft()
            # å¼¹å‡ºå³ä¾§ä¸ç¬¦åˆè¦æ±‚çš„ï¼Œä¸€ä¸ªå³å¯
            while window and nums[i] > nums[window[-1]]:
                window.pop()
            # æ›´æ–°ç°åœ¨çª—å£
            window.append(i)
            # å¦‚æœè¶…è¿‡äº†kï¼Œå°±å¼€å§‹è®°å½•ï¼Œä¹‹å‰ä¸éœ€è¦è®°å½•ã€‚å®é™…è¿”å›ç»“æœä¹Ÿæ¯”åŸelementæ•°é‡æ›´å°‘ï¼Œæ˜¯len(nums)-k+1ä¸ªç»“æœ
            if i >= k-1:
                res.append(nums[window[0]])
        return res
# Test cases
def test_maxSlidingWindow():
    s = Solution()
    assert s.maxSlidingWindow([1,3,-1,-3,5,3,6,7], 3) == [3,3,5,5,6,7]
    assert s.maxSlidingWindow([1], 1) == [1]
    assert s.maxSlidingWindow([9,11], 2) == [11]
    print("All test cases passed!")

test_maxSlidingWindow()
```

## 15. 3Sum (Medium)

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] 
such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

**Example:**
- Input: nums = [-1,0,1,2,-1,-4]
- Output: [[-1,-1,2],[-1,0,1]]


```python
from typing import List
# æ‰€æœ‰æ’åºéƒ½èƒ½æ˜¾è‘—å‡å°‘listæ‰¾æ•°å­—çš„å¤§O
# å…ˆå›ºå®šä¸€ä¸ªï¼ˆ4sumæ˜¯ä¸¤ä¸ªï¼‰ç„¶åå†åŒæŒ‡é’ˆ
# å¤–å±‚æ£€æŸ¥å¤§äº0æˆ–å°äº0
# ä¸€æ—¦ç­‰äº0ï¼Œå­˜å‚¨tripetï¼Œåˆå› ä¸ºé¢˜ç›®è¦æ±‚unique tripetï¼Œè¿™é‡Œè¦é¢å¤–å¢åŠ ä¸€ä¸ªå»é‡è®¾å®šï¼ˆå»é‡ç”¨==æ¥æ§åˆ¶è€Œä¸æ˜¯ï¼=ï¼‰
# while åµŒå¥— whileçš„æ—¶å€™ï¼Œæ³¨æ„å¤–å±‚æ¡ä»¶å’Œå†…å±‚æ¡ä»¶è¦é€‚å½“é‡å ï¼Œå¤–å±‚j<kï¼Œå†…å±‚controlæ—¶ä¹Ÿè¦å…ˆåŠ ä¸Šj<k

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        # TODO: Implement your solution
        nums = sorted(nums)
        res = []
        for i in range(len(nums)-2):
            if i>0 and nums[i] == nums[i-1]:
                continue
            j = i+1
            k = len(nums)-1
            while j < k:
                s = nums[i] + nums[j] + nums[k] 
                if s == 0:
                    res.append([nums[i], nums[j], nums[k]])
                    j += 1
                    k -= 1
                    while j < k and nums[j] == nums[j-1]:
                        j += 1
                    while j < k and nums[k] == nums[k+1]:
                        k -= 1
                elif s < 0:
                    j += 1
                else:
                    k -= 1
        return res


# Test cases
def test_threeSum():
    s = Solution()
    result = s.threeSum([-1,0,1,2,-1,-4])
    result_sorted = sorted([sorted(x) for x in result])
    expected = sorted([[-1,-1,2],[-1,0,1]])
    assert result_sorted == expected
    assert s.threeSum([0,0,0]) == [[0,0,0]]
    assert s.threeSum([0,1,1]) == []
    print("All test cases passed!")

test_threeSum()
```
